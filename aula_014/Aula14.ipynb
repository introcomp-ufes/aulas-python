### 1.1. O que  uma estrutura?

Uma Estrutura de Dados consiste em uma organizao de dados na memria de um computador ou em um dispositivo de armazenamento, de modo que esses dados possam ser utilizados de forma eficiente.

Estruturas de dados adequadas nos permitem administrar uma grande quantidade de dados de forma eficiente. Por conta disso, essas estruturas so aplicadas em bancos de dados ou servios de busca e indexao de dados, e no desenvolvimento de algoritmos eficientes.

## 2. Matrizes

Python possui suporte "nativo" a "matrizes", mas isso no funciona muito bem. O suporte no  nativo pois, na verdade, as matrizes so definidas como listas de listas. Os problemas surgem quando queremos multiplicar uma matriz por um vetor, ou uma matriz por outra matriz. Nestes casos, dois problemas principais podem ocorrer: performance e praticidade. A performance  afetada pelo uso excessivo de loops; quanto  praticidade, no  nada amigvel multiplicar "matrizes" (ou seja, listas de listas) em Python.

Para contornar estes problemas, utilizaremos a biblioteca `numpy` que oferece inmeras ferramentas matemticas, inclusive, oferece apoio ao desenvolvimento de matrizes com sintaxe bastante amigvel. `numpy` ainda  otimizada para operaes com matrizes e, portanto, performance no ser mais um problema.

Para aqueles que no conhecem um vetor, iremos defin-lo como algo muito semelhante a uma lista. Pense num vetor $A = \begin{bmatrix} 1 & 2 & 3 & 4 \end{bmatrix}$. Muito semelhantes s listas que j conhecemos, certo? Se quisermos acessar o elemento de valor $2$ devemos acessar o segundo ndice do vetor, ou seja, `A[1]`, exatamente como fazemos com listas.

Uma matriz seria um vetor com vrios vetores. Imagine agora o seguinte vetor:

$M = \begin{bmatrix} M_{[0]} & M_{[1]} & M_{[2]} & M_{[3]} \end{bmatrix}$

Se `M[0]`, `M[1]`, `M[2]` e `M[3]` forem vetores, `M`  uma matriz. Imagine `M[0] = [1, 5, 9, 13]`, `M[1] = [2, 6, 10, 14]`, `M[2] = [3, 7, 11, 15]` e `M[3] = [4, 8, 12, 16]`. Temos, nesse caso:

$M = \begin{bmatrix} 1 & 5 & 9 & 13 \\
                     2 & 6 & 10 & 14 \\
                     3 & 7 & 11 & 15 \\
                     4 & 8 & 12 & 16
    \end{bmatrix}$
    
Se estivermos interessados no elemento que vale $7$ devemos utilizar a notao `M[2][1]` pois queremos o segundo elemento do terceiro vetor. De uma forma geral, temos a notao `M[i][j]` onde `i`  a linha e `j`, a coluna.
   
**Observao:** durante o ensino mdio  ensinado que o primeiro ndice de uma matriz  $1$, aqui, entretanto, utilizaremos o primeiro ndice como $0$ para simplificar o aprendizado da biblioteca.

O primeiro passo para usar tal biblioteca  import-la:

### 2.2. Multiplicando matrizes
Quando falamos de multiplicao de matrizes/vetores em `numpy`, ao menos trs definies devem vir  mente: **produto vetorial**, **produto escalar/produto matricial** e **produto elemento-a-elemento**.

#### Produto matricial:
Nada mais  do que o produto de duas matrizes;  o produto que aprendemos a fazer durante o ensino mdio. Para realizar o produto matricial devemos ter duas matrizes $A_{n{\times}m}$ e $B_{m{\times}p}$, onde $\{n, m, p\} \in \mathbb{Z}$. Tomando como exemplo duas matrizes $A_{2{\times}3}$ e $B_{3{\times}2}$:

<center>
    $A_{2{\times}3} \cdot B_{3{\times}2} = M_{2{\times}2}$<br><br>
    $A = \begin{bmatrix}
       a & b & c\\
       d & e & f\\
    \end{bmatrix}$ <br><br>
    $B = \begin{bmatrix}
       u & v\\
       w & x\\
       y & z\\
    \end{bmatrix}$ <br><br>
    $M = \begin{bmatrix}
       a \cdot u + b \cdot w + c \cdot y & a \cdot v + b \cdot x + c \cdot z\\
       d \cdot u + e \cdot w + f \cdot y & d \cdot v + e \cdot x + f \cdot z\\
    \end{bmatrix}$
</center>

Em `numpy` essa operao entre as matrizes $A$ e $B$ pode ser obtido por meio dos comandos `np.dot(A,B)` ou `A.dot(B)` [[ref]](https://numpy.org/doc/stable/reference/generated/numpy.dot.html).

````python
A = np.array([[1, 2, 3],[4, 5, 6]], dtype=int)
B = np.array([[21, 22], [23, 24], [25, 26]], dtype=int)

print(f"A = \n {A}\n")
print(f"B = \n {B}\n")
print(f"Usando A.dot:\n{A.dot(B)}\n")
print(f"Usando np.dot:\n{np.dot(A, B)}\n")
````

#### Produto escalar:
Um produto matricial que resulta em escalar (ou nmero real)  chamado produto escalar. Pode ser obtido da mesma maneira que o produto matricial. Ao fim da operao pode ser realizado _casting_ para os tipos `float` ou `int`.

````python
A = np.array([1, 2, 3], dtype=int)
B = np.array([[21], [22], [23]], dtype=int)

print(f"A = \n {A}\n")
print(f"B = \n {B}\n")
print(f"Usando A.dot:\n{A.dot(B)}\n")
print(f"Usando np.dot:\n{np.dot(A, B)}\n")
````

#### Produto elemento-a-elemento:
Abordagem diferente da encontrada no ensino mdio, devemos ter duas matrizes $A_{n{\times}m}$ e $B_{n{\times}m}$ que resultaram numa terceira matriz $C_{n{\times}m}$ onde `C[i,j] = A[i,j] * B[i,j]`, lembrando que $\{i, j, n, m\} \in \mathbb{Z}$. Para $A_{2{\times}2} \cdot B_{2{\times}2} = C_{2{\times}2}$:

<center>
    $A = \begin{bmatrix}
       a & b\\
       c & d\\
    \end{bmatrix}$ <br><br>
    $B = \begin{bmatrix}
       u & v\\
       w & x\\
    \end{bmatrix}$ <br><br>
    $C = \begin{bmatrix}
       a \cdot u & b \cdot v\\
       c \cdot w & d \cdot x\\
    \end{bmatrix}$
</center>

Em `numpy` essa  a abordagem mais simples, basta chamar `A * B`.

````python
A = np.array([[1, 2, 3],[4, 5, 6]], dtype=int)
B = np.array([[21, 22, 23], [24, 25, 26]], dtype=int)

print(f"A = \n {A}\n")
print(f"B = \n {B}\n")
print(f"Usando A*B:\n{A*B}\n")
````

#### Diviso elemento-a-elemento:
Semelhante ao produto elemento-a-elemento, possui inclusive os mesmos requisitos. Ao invs de realizar os produtos dos elementos de $A$ e $B$  realizada diviso. Basta chamar `A/B`.

#### Soma de matriz com vetor:
Ao invs de ter que utilizar _loops_ ou outros artifcios para transformar um vetor em matriz e assim poder som-lo com outra matriz, em `numpy` podemos simplesmente somar um vetor $V_{n{\times}1}$ e uma matriz $M_{n{\times}m}$ que a biblioteca se encarregar de resolver todos os problemas.

````python
V = np.array([1, 2], dtype=int)
M = np.array([[2, 1], [2, 1]], dtype=int)

print(V + M)
````

#### Matriz transposta:
Transpr uma matriz/vetor  muitas vezes essencial, imagine que voc possui $A_{n{\times}m}$ e $B_{n{\times}m}$ e quer realizar seu produto matricial. Com a transposta de $A$ ou $B$ isso  plenamente possvel. Em `numpy` a transposta de $A$  dada por `A.T`.

````python
A = np.array([[1, 2, 3], [4, 5, 6]])
B = np.array([[7, 8, 9], [1, 3, 5]])

print(A.dot(B.T))

V = np.array([[0, 1, 1],[3, 5, 8]])

print(V)
print(V.T)
````

#### Filtrar contudo de vetor:
Suponha que possua um vetor de nmeros $A$, seus nmeros so aleatrios e voc s possui interesse naqueles que sao positivos. Como faria para filtrar? Em `numpy` isso  extremamente simples. Primeiro, devemos entender que podemos escolher quais ndices de vetor vamos querer exportar para outro vetor.

````python
np.random.seed(3)
A = np.random.randn(20, 1)

print(f"Primeiro nmero de A: {A[0]}")
print(f"Quinto nmero de A: {A[4]}\n")
print(f"Vetor com os dois nmeros:\n{A[[0, 4]]}")
````

Perceba que voc pode pedir vrios elementos ao mesmo tempo passando uma lista como ndice. Para que possamos excluir os nmeros negativos, basta descobrirmos quais ndices possuem nmeros positivos. Para tanto podemos utilizar a propriedade de comparao de `numpy`.

````python
A >= 0
````

Perceba que ao pedir `A > 0`,  retornado um `np.ndarray` dizendo se naquele ndice $i$, o valor $A_{[i]}$,  maior ou igual a 0. Portanto no precisamos saber quais ndices possuem nmeros negativos, podemos simplesmente passar esse `np.ndarray` informando quais ndices temos interesse. No temos interesses nos ndices marcados com `False`, apenas os que possuem `True`.

````python
A_filtrado = A[A >= 0]

print(f"A =\n{A.T}\n")
print(f"A_filtrado =\n{A_filtrado}")
````

#### Demais funes em vetores:
Podemos utilizar a mesma propriedade da comparao para realizar operaes rapidamente em um vetor inteiro. Podemos, por exemplo, aplicar a funo $sen({\theta})$ em todo um vetor.

````python
X = np.arange(-np.pi, np.pi, 0.01) # np.arange  semelhante  funo range, mas retorna um np.ndarray.
# Podemos definir o tamanho dos "passos" em np.arange ao invs de utilizar o padro 1 como em range().
Y = np.sin(X) # np.sin aplica a funo seno em todos os ndices do vetor

print(np.max(Y)) # np.max retorna o maior valor no vetor
````

A estrutura de dados conjunto possui trs propriedades importantes:  desordenada,  imutvel e no permite duplicatas. Exploraremos cada uma dessas propriedades.

````python
meu_conjunto = {1, 2, 3, -1, -10}
meu_conjunto
````

### 3.3. Operaes com conjuntos

Complementando o que foi dito acima, existe toda uma **lgebra de conjuntos** disponvel para que possamos nos divertir. Boa parte dos **Teoremas de Probabilidade** utilizam conjuntos. O nvel de abstrao dessa estrutura permite resolver facilmente problemas de programao competitiva.

A primeira operao  a **unio**, como o prprio nome sugere, ela permite a juno dos elementos de dois conjuntos. A unio  representada pelo smbolo `|` ou pela chamada `conjunto1.union(conjunto2)`.

````python
frutas = {"Banana", "Uva", "Abacaxi", "Tomate"}
pecas_automotivas = {"Motor V8", "Graxa", "Escapamento", "Carburador"}
supermercado = {"Banana", "Uva", "Abacaxi", "Tomate", "Cereal", "Biscoito", "Leite"}
lista_de_compras = frutas | pecas_automotivas

check = all(item in supermercado for item in lista_de_compras)

if check == True:
    print("Oba! Consegui comprar meu carburador no supermercado.")
else:
    print("Poxa, parece que precisarei procurar uma oficina mecnica.")
````

````python
lista_de_compras
````

Perceba que, na situao do cdigo acima, se nossa lista de compras fosse composta apenas de frutas, conseguiramos comprar tudo. A operao de unio mesclou os itens automotivos com as frutas, fazendo com que no fosse possvel encontrar todos os itens no supermercado.

Outra operao comum em conjuntos  a interseo, que nos retorna os elementos que aparecem em ambos conjuntos. O exemplo abaixo ilustra essa operao.

````python
conjunto1 = {0, 1, 3, 5, 7, 9}
conjunto2 = {0, 5, 10, 15, 20}
conjunto3 = conjunto1 & conjunto2 
print(conjunto3)
````

Ou ento podemos fazer assim:

````python
conjunto4 = conjunto1.intersection(conjunto2)
print(conjunto4)
````

Outra coisa bem legal que podemos fazer  a diferena entre dois conjuntos. Podemos usar a operao difference() para retornar a diferena. Por exemplo, se temos um conjunto A e um conjunto B, a diferena ser A - B, que so os elementos de A que no esto em B. Veja um exemplo:

````python
conjunto5 = conjunto1.difference(conjunto2)
print(conjunto5)
````

Outra jeito de fazer isso :

````python
conjunto6 = conjunto1 - conjunto2
print(conjunto6)
````

Dicionrios (a estrutura de dados que estamos estudando, no o livro) no se limitam  associao de uma palavra  um significado. Podemos associar _strings_  nmeros, nmeros  _strings_ e quaisquer demais combinaes que conseguirmos imaginar e que nos forem convenientes no momento.

### 4.1. Declarando um dicionrio

O cdigo a seguir mostra uma das maneiras de se declarar um dicionrio, seguindo a sintaxe

 importante salientar que dicionrios so mutveis, ou seja, podemos adicionar/modificar chaves/valores aps a declarao de um dicionrio.

dicionrio[chave] = valor

O dicionrio pode armazenar diversos tipos de dados, no s os tipos de dados primitivos (int, float, etc.). Podemos incluir listas, tuplas e vrios outros tipos de dados ao nosso dicionrio, at mesmo um outro dicionrio!

````python
filhos = [pessoa1]
pessoa2['filhos'] = filhos
print(pessoa2)
````

#### Alterando valores 
Como dito acima, dicionrios so mutveis. Com isso, sabemos que valores j definidos em dicionrios podem ser alterados (e novos podem ser inseridos).
