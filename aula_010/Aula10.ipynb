## 1. Paradigmas de Programação

Antes de começar o estudo do paradigma de programação orientada a objetos (POO), é preciso entender que existem vários outros paradigmas. Paradigma é um meio de se classificar uma linguagem de acordo com suas funcionalidades. Entre os paradigmas existentes, podemos citar o procedural, o funcional, o estruturado, o imperativo e o orientado a objetos.

No mundo do desenvolvimento de softwares existem as linguagens conhecidas como multiparadigmas. Como o nome sugere, essas linguagens suportam mais de um paradigma, como é o caso de Python. Um dos paradigmas suportados por Python é o de POO (Programação Orientada a Objetos), o qual abordamos nesta aula. Python também permite o uso de Programação Funcional, mas isso não será tratado nessa aula. É importante notar que, até agora no curso, utilizamos apenas o paradigma imperativo/procedural.

### 1.1. Programação Imperativa

O paradigma de programação imperativa é fortemente baseada no modelo apresentado pelo matemático ___Alan Turing___, denominado "A máquina de Turing", no qual o programador especifica passos que devem ser seguidos em ordem pelo computador.

Quase sempre, em conjunto com a programação imperativa, utilizamos o paradigma de __programação procedural__, ou seja, orientada a procedimentos. Um procedimento nada mais é do uma função que criamos e utilizamos.

Esses são os paradigmas que mais utilizamos durante o curso, e os únicos que utilizamos até agora. O trecho de código a seguir ilustra um exemplo simples de programação imperativa para calcular a distância percorrida por um carro.

````python
# Calculando a distância percorrida por um carro utilizando programação imperativa:

# DADOS:
velocidade_do_carro = 60 #km/h
tempo_de_viagem = 140 # min

# Processando os dados para obter um resultado:
horas_de_viagem = tempo_de_viagem / 60 #h
distancia_percorrida = velocidade_do_carro * horas_de_viagem #km

# Exibindo a saída:
print(f"{distancia_percorrida:.0f}km")
````

### 1.3. Outros paradigmas

Além da programação imperativa, procedural e orientada a objetos existem muitos outros paradigmas que podem ser explorados. Um exemplo é a programação funcional, que pode fazer coisas incríveis em pouquíssimas linhas. Daremos apenas um exemplo, não se preocupe em entender.

````python
# Criando uma lista de números pares com programação imperativa:
pares = []
for i in range(1, 101):
    if i % 2 == 0:
        pares.append(i)
        
print(pares)
````

````python
# Criando uma lista de números pares com programação funcional:
pares = [ i for i in range(1, 101) if i % 2 == 0 ]
print(pares)
````

Como podemos observar, com programação funcional é possível reduzir quatro linhas de código em apenas uma. A programação funcional cresce a cada dia e é um excelente aprendizado caso queira entrar no mundo de Go, Haskell e Kotlin, linguagens utilizadas na análise de dados e no mercado _mobile_.

## 2. Uma visão mais profunda de POO

Em POO, tentamos aplicar conceitos de objetos reais na programação. Ora, um carro possui suas características, certo? Além disso, carros executam/realizam ações.

Por exemplo, podemos considerar as seguintes características de um carro:

* Ano de fabricação
* Proprietário
* Dimensões (Altura, Largura)
* Velocidade Máxima
* Aceleração

Podemos também pensar que o carro realiza as seguintes ações:

* Acelerar
* Ligar/desligar o motor
* Ligar/desligar o rádio

Agora, imagine que você precise "criar" um carro em Python. Podemos considerar as seguintes alternativas:

#### 1. Criar as funções separadamente:

Bem, você poderia simplesmente criar as variáveis `carro_ano_de_fabr`, `carro_proprietario` e todo o resto. Poderia também criar a função `acelerar_carro(carro_velocidade_maxima, carro_aceleracao, tempo)` caso quisesse implementar a aceleração em seu carro. Isso aparenta ser um trabalho muito grande, não é mesmo? Principalmente quando estamos falando de centenas de carros diferentes.

#### 2. Utilizar um dicionário:

Com um dicionário, poderíamos simplesmente fazer:

``` python
carro = {
    "ano_de_fabr" : 2010,
    "proprietario" : "João",
    .
    .
    .
}

```

E isso facilitaria muito nosso trabalho! Poderíamos até mesmo automatizar o processo e criar quantos carros quisermos. Para acelerar o carro basta chamar a função `acelera_carro(carro)`. Essa solução já é mais simples que a do exemplo anterior.

#### 3. Criar uma classe:

Se utilizássemos POO tudo ficaria ainda mais elegante. Poderíamos criar um objeto carro apenas chamando um **construtor** que "configura" o nosso carro:

``` python
carro = Carro(ano_de_fabr, proprieterio, ..., aceleracao)
```

e chamar o **método** acelerar da seguinte forma:

``` python
carro.acelerar()
```

Muito melhor, certo? O uso de objetos também permite que façamos tratamento de dados de forma mais eficiente e organizada, o que evita erros.

**Obs.:** Um objeto é uma instância de uma classe. É como dizer que "Joãozinho", uma pessoa específica, pertence à classe "Humanos". Nesse caso "Joãozinho" é um objeto.

### 2.1. Abstração

Lembre-se que queremos trabalhar com conceitos reais de objetos, então precisamos estudar e entender o que tal representação de um objeto real realizará dentro do nosso sistema.

Nosso carro é bom um exemplo de abstração, pois nele nos concentramos nas propriedades e ações que são de interesse para nosso problema. Se estivéssemos implementando um sistema para uma fábrica de carros, teríamos outras propriedades e ações de interesse (exemplo: material, peso, custo, etc.). Isso é a abstração, você deixa de trabalhar com todas as variáveis e se preocupar apenas com as características relevantes do seu **objeto**.

Em orientação a objetos, as características do objeto (por exemplo, o carro) são chamadas **propriedades**, e suas ações são chamadas **métodos**. Portanto temos, como exemplos, a propriedade `carro.proprietario` e o método `carro.acelerar()`.

### 2.2. Encapsulamento

Imagine uma classe `Soma` e que essa classe seja composta de 3 propriedades e 2 métodos, como a seguir:

```
    classe Soma:
        Propriedades:
            Elemento1,
            Elemento2,
            Resultado
        Métodos:
            Soma(Elemento1, Elemento2), # Construtor
            exibirOperação()
```

Se quisermos representar a soma $2 + 3$ basta executarmos o **construtor** "Soma", como a seguir:

`soma = Soma(2, 3)`

O construtor se encarregará de atribuir à propriedade "Resultado" o valor 5 (aprenderemos mais sobre construtores posteriormente). Entretanto nada impede que o usuário de nossa classe faça algo como `soma.Resultado = 8` e, ao chamar `soma.exibirOperação()` teríamos na saída algo como `2 + 3 = 8`, o que é um absurdo. Como podemos garantir que esse tipo de erro não ocorra?

O encapsulamento de nossas classes está relacionado justamente a isso, evitar modificações indevidas e manter um determinado nível de abstração (uma vez que podemos "esconder" as entranhas de nossa classe, exibindo apenas aquilo que é interessante ao usuário).

Para resolvermos nosso problema com a classe `Soma`, precisamos entender o conceito de **modificadores de acesso**. Isso nada mais é do que uma "trava" para determinados elementos de nossa classe. Propriedades e métodos podem ser **públicos** ou **privados**. Como sugerido pela própria nomenclatura, um elemento privado só pode ser acessado pela própria classe, sendo seu uso bloqueado externamente. Um elemento público pode ser acessado e modificado livremente. Em Python todas as propriedades e todos os métodos são públicos, a não ser que você defina um modificador de acesso. Conhecendo tais conceitos, vamos atualizar a representação de nossa classe.

```
    classe Soma:
        Propriedades:
            Elemento1 (Pública),
            Elemento2 (Pública),
            Resultado (Pública)
        Métodos:
            Soma(Elemento1, Elemento2) (Público), # Construtor
            exibirOperação() (Público)
```

Logo de cara percebemos que a propriedade `Resultado` não deveria ser livre, tampouco as propriedades `Elemento1` e `Elemento2` (imagine se alterarmos os elementos de uma soma e mantermos o resultado, na maioria dos casos estaríamos "estragando" a soma). Os métodos, no entanto, devem permanecer públicos já que é de nosso interesse que o usuário possa acessá-los sempre que quiser. Isso é recorrente em muitas classes, ou seja, os métodos costumam ser públicos e as propriedades costumam ser privadas. Vamos atualizar nossa classe, agora sem especificar quando o elemento é público:

```
    classe Soma:
        Propriedades:
            Elemento1 (Privada),
            Elemento2 (Privada),
            Resultado (Privada)
        Métodos:
            Soma(Elemento1, Elemento2), # Construtor
            exibirOperação()
```

Mas e se quisermos, por exemplo, apenas realizar um simples `print(soma.Resultado)` ou então modificar os elementos corretamente? Para esse fim criaremos métodos conhecidos como _**getters**_ e _**setters**_. _Getters_ são responsáveis por retornar os valores das propriedades e _setters_ são responsáveis pela modificação dessas propriedades. Definindo nossos _getters_ e _setters_:

```
    classe Soma:
        Propriedades:
            Elemento1 (Privada),
            Elemento2 (Privada),
            Resultado (Privada)
        Métodos:
            Soma(Elemento1, Elemento2), # Construtor
            modificarSoma(Elemento1, Elemento2), # Setter (Define novos valores para Elemento1 e Elemento2)
            receberElementos(), # Retorna os valores de Elemento1 e Elemento2
            receberResultado() # Retorna o valor de Resultado
```

É importante frisar que dentro do método `modificarSoma` não estamos simplesmente modificando `Elemento1` e `Elemento2`. Estamos, na realidade, fazendo algo do tipo:

```
    método modificarSoma(NovoElemento1, NovoElemento2):
        Elemento1 = NovoElemento1
        Elemento2 = NovoElemento2
        Resultado = NovoElemento1 + NovoElemento2
```

### 2.3. Herança

Ainda em nosso exemplo da classe `Soma`, podemos definir a soma como herdeira da classe `Operação`. Mas como assim herdeira? Veja bem, existem quatro operações matemáticas básicas (adiçao, multiplicação, subtração e divisão) e todas elas possuem características em comum. As características comuns das operações matemáticas são:

* Possuem dois elementos (operandos)
* Possuem um resultado

Portanto, podemos definir uma classe mais genérica chamada `Operação` da seguinte forma:

```
    classe Operação:
        Propriedades:
            Elemento1 (Privada),
            Elemento2 (Privada),
            Resultado (Privada)
        Métodos:
            Operação(Elemento1, Elemento2),
            modificarElementos(Elemento1, Elemento2),
            receberElementos(),
            receberResultado()
```

Podemos agora recriar a classe `Soma` como **herdeira** da classe `Operação`:

```
    classe Soma (herdeira de Operação):
        Métodos:
            Soma(Elemento1, Elemento2),
            modificarElementos(Elemento1, Elemento2)
```

Perceba que não declaramos nenhuma propriedade. Com a herança, a classe `Soma` **herda** todos os métodos e propriedades da classe `Operação`. Com isso precisamos apenas **sobrescrever** os métodos que terão comportamento diferente (uma soma não funciona da mesma forma que uma divisão e, portanto, seus construtores e _setters_ são diferentes).

Para criar uma classe em Python basta utilizar a palavra-chave `class` seguida do nome da classe. No nosso caso seria algo como:



```python
class Soma:
```

Perceba o parâmetro `self`. "_self_" em inglês significa algo como "próprio" ou "auto" e em Python ele representa o próprio objeto, ou seja, modificações na variável `self` são modificações no objeto. Dessa maneira sabemos que estamos modificando as propriedades `elemento1` e `elemento2` do objeto logo na sua construção.

Vamos declarar nossa classe:

````python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operação (int ou float)
        elemento2 : segundo elemento da operação (int ou float)
        resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
````

Vamos realizar alguns testes:

````python
soma = Soma(1, 2) # Criando o objeto soma através do construtor.
````

Perceba que não precisamos escrever `Soma.__init__(self, ...)`, o interpretador entende que ao chamarmos o nome da classe com
parênteses estamos nos referindo ao construtor. O parâmetro `self` também deve ser ignorado em chamadas.

````python
# Exibindo nossa operação de forma elegante (nem tanto, mais tarde veremos como encapsular):
print(f"{soma.elemento1} + {soma.elemento2} = {soma.resultado}")
````

### 3.2. Métodos

Como dito antes, métodos são as "funções" das classes. Vamos colocar a mão na massa e criar o método `exibir_operacao`:

````python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operação (int ou float)
        elemento2 : segundo elemento da operação (int ou float)
        resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
    
    def exibe_operacao(self):
        string_operacao = f"{self.elemento1} + {self.elemento2} = {self.resultado}"
        print(string_operacao)
````

````python
soma = Soma(2, 2)
soma.exibe_operacao()
````

Agora ficou muito mais simples exibir a operação. Contudo existem maneiras melhores de implementar essa funcionalidade. A melhor maneira é criar um método chamado `__str__` que converte seu objeto em uma _string_. Assim como `__init__` e `__str__`, você pode criar alguns métodos com nomes "especiais" que irão melhorar sua classe, esses métodos podem ser encontrados [aqui](https://docs.python.org/3/library/functions.html).

````python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operação (int ou float)
        elemento2 : segundo elemento da operação (int ou float)
        resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
    
    def __str__(self):
        string_operacao = f"{self.elemento1} + {self.elemento2} = {self.resultado}"
        return string_operacao
````

````python
soma = Soma(2, 2)
print(soma)
````

### 4.1. Encapsulamento

Retomando nosso exemplo da classe `Soma` presente nas seções 2 e 3, você já deve ter percebido que nossa classe `Soma` ainda não possui nenhum tipo de encapsulamento e, portanto, podemos alterar seus parâmetros livremente. Nossa primeira tarefa é definir os `getters` e `setters`. Note que já definimos um setter (o método `modificar_elementos`), precisamos agora definir os getters.

Uma coisa importante que precisamos definir é como chamaremos esses métodos, existem duas formas de declarar (e posteriormente chamar) _getters_ e *setters*, observe a tabela abaixo:

<table>
    <tr>
        <th></th>
        <th>Jeito não pythônico</th>
        <th>Jeito pythônico</th>
    </tr>
    <tr>
        <td>GET</td>
        <td>get_parametro()</td>
        <td>parametro</td>
    </tr>
    <tr>
        <td>SET</td>
        <td>set_parametro(valor)</td>
        <td>parametro=valor</td>
    </tr>
</table>

Primeiro definiremos do jeito não-_pythônico_ (é mais simples, mas não tão elegante):

```python
class Soma:
    [...]
    def get_elemento1(self):
        return self.elemento1
        
    def get_elemento2(self):
        return self.elemento2
        
    def get_resultado(self):
        return self.resultado
```

O jeito _pythônico_ de se fazer isso é utilizando um mecanismo da linguagem chamado ___decorator___, que auxilia o interpretador e pode ser muito útil em seu código. Não se preocupe se não conseguir entender, fazer as coisas do jeito não-_pythônico_ não tornará seu código inutilizável.

Para definir os _getters_ utilizamos o _decorator_ `@property` uma linha antes de definir o método.

```python
class Soma:
    [...]
    @property
    def elemento1(self):
        return self.elemento1
        
    @property
    def elemento2(self):
        return self.elemento2
        
    @property
    def resultado(self):
        return self.resultado
```

Para os _setters_ utilizamos o _decorator_ `@nome_da_propriedade.setter`:

```python
class Soma:
    [...]
    @elemento1.setter
    def elemento1(self, valor):
        self.elemento1 = valor
        self.resultado = valor + self.elemento2
        
    @elemento2.setter
    def elemento2(self, valor):
        self.elemento2 = valor
        self.resultado = self.elemento2 + valor
```

Desse modo não precisamos mais do método `modificar_elementos`. Vamos ver como ficou nossa classe:

```python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operação (int ou float)
        elemento2 : segundo elemento da operação (int ou float)
        resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
    
    def __str__(self):
        string_operacao = f"{self.elemento1} + {self.elemento2} = {self.resultado}"
        return string_operacao

    @property
    def elemento1(self):
        return self.elemento1

    @property
    def elemento2(self):
        return self.elemento2

    @property
    def resultado(self):
        return self.resultado
    
    @elemento1.setter
    def elemento1(self, valor):
        self.elemento1 = valor
        self.resultado = valor + self.elemento2

    @elemento2.setter
    def elemento2(self, valor):
        self.elemento2 = valor
        self.resultado = self.elemento1 + valor
```

Ainda não podemos usar os recém criados _getters_ e _setters_ (o interpretador apontará erros). Precisamos definir nossas propriedades como privadas. Para isso devemos inserir um _underscore_ duplo (`__`) antes do nome de nossa propriedade. Dessa maneira:

```python
class Soma:
    """
    Propriedades:
        __elemento1 : primeiro elemento da operação (int ou float)
        __elemento2 : segundo elemento da operação (int ou float)
        __resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.__elemento1 = elemento1
        self.__elemento2 = elemento2
        self.__resultado = elemento1 + elemento2
    
    def __str__(self):
        string_operacao = f"{self.__elemento1} + {self.__elemento2} = {self.__resultado}"
        return string_operacao

    @property
    def elemento1(self):
        return self.__elemento1

    @property
    def elemento2(self):
        return self.__elemento2

    @property
    def resultado(self):
        return self.__resultado
    
    @elemento1.setter
    def elemento1(self, valor):
        self.__elemento1 = valor
        self.__resultado = valor + self.__elemento2

    @elemento2.setter
    def elemento2(self, valor):
        self.__elemento2 = valor
        self.__resultado = self.__elemento1 + valor
```

Vamos colocar a classe para funcionar:

````python
class Soma:
    """
    Propriedades:
        __elemento1 : primeiro elemento da operação (int ou float)
        __elemento2 : segundo elemento da operação (int ou float)
        __resultado : resultado da operação (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.__elemento1 = elemento1
        self.__elemento2 = elemento2
        self.__resultado = elemento1 + elemento2

    def __str__(self):
        string_operacao = f"{self.__elemento1} + {self.__elemento2} = {self.__resultado}"
        return string_operacao

    @property
    def elemento1(self):
        return self.__elemento1

    @property
    def elemento2(self):
        return self.__elemento2

    @property
    def resultado(self):
        return self.__resultado

    @elemento1.setter
    def elemento1(self, valor):
        self.__elemento1 = valor
        self.__resultado = valor + self.__elemento2

    @elemento2.setter
    def elemento2(self, valor):
        self.__elemento2 = valor
        self.__resultado = self.__elemento1 + valor
````

````python
soma = Soma(2, 2)
print(soma)

soma.elemento1 = 3
print(soma)

soma.elemento2 = 4
print(soma)

print(soma.elemento1)
````

Perceba como tudo funciona naturalmente do jeito _pythônico_ (parece até que estamos atribuindo os valores diretamente às propriedades mas, no entanto, estamos utilizando os setters). Se fizéssemos do jeito mais tradicional, teríamos algo como:

```python
soma = Soma(2, 2)

soma.set_elemento1(3)
print(soma)

print(soma)

```

### 4.2. Herança

Para uma perspectiva prática de herança iremos criar a classe `Operacao` que realiza operações matemáticas.
```
classe Operacao:
    Propriedades:
        num1 : (inteiro) primeiro número da operação
        oper : (caractere) operador, define a operação a ser realizada
        num2 : (inteiro) segundo número da operação
    Métodos:
        resultado : retorna o resultado da operação
```

````python
class Operacao:
    """
    Propriedades:
        num1 : (int) primeiro número da operação
        oper : (str) operador, define a operação a ser realizada
        num2 : (int) segundo número da operação
    """
    def __init__(self, num1, oper, num2):
        self.num1 = num1
        self.oper = oper
        self.num2 = num2
        
    def resultado(self):
        if self.oper == "+":
            return self.num1 + self.num2
        elif self.oper == "-":
            return self.num1 - self.num2
        elif self.oper == "*":
            return self.num1 * self.num2
        elif self.oper == "/":
            return self.num1 / self.num2
````

````python
# Célula testadora
oper = Operacao(2, "*", 2) # 2 x 2
oper.resultado() # Deve retornar 4
````

Temos uma classe genérica de operações matemáticas, podemos criar a classe `Soma` a partir dessa classe. Basta herdarmos ela quando implementarmos `Soma`.

````python
class Soma(Operacao): # Definimos Soma como filha (ou herdeira) de Operacao
    # Ao definirmos isso, Soma pode acessar métodos e propriedades privadas de Operacao
    # Soma também herda métodos e propriedades públicas
    """
    Propriedades:
        num1 : (int) primeiro número da soma
        num2 : (int) segundo número da soma
    """
    def __init__(self, num1, num2):
        super().__init__(num1, "+", num2) # Referência à classe mãe (aquela da qual Soma herda suas características)
````

````python
oper = Soma(3, 2) # 3 + 2
oper.resultado() # Deve retornar 5
````

Perceba como é simples, não precisamos definir o método `resultado` em `Soma` e seu funcionamento foi excelente, uma vez que o objeto gerou o mesmo resultado que `Operacao(num1, "+", num2)`, e foi exatamente isso que criamos. A palavra chave `super` referenciou a classe mãe, e podemos provar isso implementando de outra forma:

````python
class Soma(Operacao):
    """
    Propriedades:
        num1 : (int) primeiro número da soma
        num2 : (int) segundo número da soma
    """
    def __init__(self, num1, num2):
        Operacao.__init__(self, num1, "+", num2)
````

````python
oper = Soma(3, 2) # 3 + 2
oper.resultado() # Deve retornar 5
````

Desse modo podemos definir um modelo para uma classe herdeira:
```python
class <NomeDaClasseHerdeira>(<NomeDaClasseMae>):
    def __init__(self, <argumentos>):
        super().__init__(<argumentos>)
```
Logicamente esse modelo irá variar dependendo de suas necessidades, não existe um padrão, esse é apenas o mais simplista. Lembre-se que uma classe herdeira nem sempre é mais simples do que sua classe mãe, muitas das vezes criamos classes mães extremamente simples (chamadas interfaces) e implementamos suas herdeiras de maneira muito mais completa. É um conceito que requer um estudo muito mais aprofundado e que pode ser combinado com **polimorfismo** para gerar códigos muito elegantes e manuteníveis.

## 5. O que pode dar errado?

A classe abaixo representa um triângulo retângulo. Como bem sabemos um triângulo retângulo de catetos com, respectivamente, $a$ e $b$ de comprimento e hipotenusa de $c$ de comprimento deve seguir a fórmula:
$c^2 = a^2 + b^2$

````python
import math
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """
    def __init__(self, cateto_a, cateto_b):
        self.cateto_a = cateto_a
        self.cateto_b = cateto_b
        self.hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    @property
    def cateto_a(self):
        return self.cateto_a
    
    @property
    def cateto_b(self):
        return self.cateto_b
    
    @property
    def hipotenusa(self):
        return self.hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.cateto_a = valor
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.cateto_b = valor
        
TrianguloRetangulo(3, 4)
````

Observe que obtemos uma mensagem de erro do tipo `AttributeError`. Ela nos indica que estamos tentando definir o valor de um atributo sem que o seu `setter` tenha sido definido apropriadamente. Isso ocorreu pois estamos atribuindo valor à propriedade declarada com o decorator `@property` e não à propriedade privada, que nem mesmo chegou a ser declarada. Podemos resolver este problema da seguinte forma:

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    @property
    def cateto_a(self):
        return self.__cateto_a
    
    @property
    def cateto_b(self):
        return self.__cateto_b
    
    @property
    def hipotenusa(self):
        return self.__hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.__cateto_a = valor
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.__cateto_b = valor
````

````python
# Célula exemplo:
tri = TrianguloRetangulo(3, 4)
print(tri)
tri.cateto_a = 10
print(tri)
````

Perceba que alteramos o valor do cateto A e ainda assim a hipotenusa manteve seu valor original. Isso é um erro comum ao lidar com objetos e entra na categoria dos erros de semântica, você não será notificado pelo interpretador e ele irá aparecer na pior hora possível. Devemos atualizar o valor da hipotenusa sempre que alterarmos o valor de `cateto_a` e `cateto_b`.

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__atualiza_hipotenusa()
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    def __atualiza_hipotenusa(self):
        self.__hipotenusa = (self.cateto_a**2 + self.cateto_b**2)**(1/2)
    
    @property
    def cateto_a(self):
        return self.__cateto_a
    
    @property
    def cateto_b(self):
        return self.__cateto_b
    
    @property
    def hipotenusa(self):
        return self.__hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.__cateto_a = valor
        self.__atualiza_hipotenusa()
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.__cateto_b = valor
        self.__atualiza_hipotenusa()
````

````python
# Célula exemplo:
tri = TrianguloRetangulo(3, 4)
print(tri)
tri.cateto_a = 6
tri.cateto_b = 8
print(tri)
````

Agora tudo parece funcionar corretamente. O segredo para a solução de problemas relacionados a classes é sempre ficar atento e lembrar que o depurador é seu melhor amigo. Isso se deve ao fato de que, em orientação a objetos, erros de semântica tendem a ser mais recorrentes, e mais graves, do que erros de sintaxe.
