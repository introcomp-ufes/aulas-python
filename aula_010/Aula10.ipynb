### 1.1. Programao Imperativa

O paradigma de programao imperativa  fortemente baseada no modelo apresentado pelo matemtico ___Alan Turing___, denominado "A mquina de Turing", no qual o programador especifica passos que devem ser seguidos em ordem pelo computador.

Quase sempre, em conjunto com a programao imperativa, utilizamos o paradigma de __programao procedural__, ou seja, orientada a procedimentos. Um procedimento nada mais  do uma funo que criamos e utilizamos.

Esses so os paradigmas que mais utilizamos durante o curso, e os nicos que utilizamos at agora. O trecho de cdigo a seguir ilustra um exemplo simples de programao imperativa para calcular a distncia percorrida por um carro.

````python
# Calculando a distncia percorrida por um carro utilizando programao imperativa:

# DADOS:
velocidade_do_carro = 60 #km/h
tempo_de_viagem = 140 # min

# Processando os dados para obter um resultado:
horas_de_viagem = tempo_de_viagem / 60 #h
distancia_percorrida = velocidade_do_carro * horas_de_viagem #km

# Exibindo a sada:
print(f"{distancia_percorrida:.0f}km")
````

### 1.3. Outros paradigmas

Alm da programao imperativa, procedural e orientada a objetos existem muitos outros paradigmas que podem ser explorados. Um exemplo  a programao funcional, que pode fazer coisas incrveis em pouqussimas linhas. Daremos apenas um exemplo, no se preocupe em entender.

````python
# Criando uma lista de nmeros pares com programao imperativa:
pares = []
for i in range(1, 101):
    if i % 2 == 0:
        pares.append(i)
        
print(pares)
````

````python
# Criando uma lista de nmeros pares com programao funcional:
pares = [ i for i in range(1, 101) if i % 2 == 0 ]
print(pares)
````

### 2.1. Abstrao

Lembre-se que queremos trabalhar com conceitos reais de objetos, ento precisamos estudar e entender o que tal representao de um objeto real realizar dentro do nosso sistema.

Nosso carro  bom um exemplo de abstrao, pois nele nos concentramos nas propriedades e aes que so de interesse para nosso problema. Se estivssemos implementando um sistema para uma fbrica de carros, teramos outras propriedades e aes de interesse (exemplo: material, peso, custo, etc.). Isso  a abstrao, voc deixa de trabalhar com todas as variveis e se preocupar apenas com as caractersticas relevantes do seu **objeto**.

Em orientao a objetos, as caractersticas do objeto (por exemplo, o carro) so chamadas **propriedades**, e suas aes so chamadas **mtodos**. Portanto temos, como exemplos, a propriedade `carro.proprietario` e o mtodo `carro.acelerar()`.

### 2.2. Encapsulamento

Imagine uma classe `Soma` e que essa classe seja composta de 3 propriedades e 2 mtodos, como a seguir:

```
    classe Soma:
        Propriedades:
            Elemento1,
            Elemento2,
            Resultado
        Mtodos:
            Soma(Elemento1, Elemento2), # Construtor
            exibirOperao()
```

Se quisermos representar a soma $2 + 3$ basta executarmos o **construtor** "Soma", como a seguir:

`soma = Soma(2, 3)`

O construtor se encarregar de atribuir  propriedade "Resultado" o valor 5 (aprenderemos mais sobre construtores posteriormente). Entretanto nada impede que o usurio de nossa classe faa algo como `soma.Resultado = 8` e, ao chamar `soma.exibirOperao()` teramos na sada algo como `2 + 3 = 8`, o que  um absurdo. Como podemos garantir que esse tipo de erro no ocorra?

O encapsulamento de nossas classes est relacionado justamente a isso, evitar modificaes indevidas e manter um determinado nvel de abstrao (uma vez que podemos "esconder" as entranhas de nossa classe, exibindo apenas aquilo que  interessante ao usurio).

Para resolvermos nosso problema com a classe `Soma`, precisamos entender o conceito de **modificadores de acesso**. Isso nada mais  do que uma "trava" para determinados elementos de nossa classe. Propriedades e mtodos podem ser **pblicos** ou **privados**. Como sugerido pela prpria nomenclatura, um elemento privado s pode ser acessado pela prpria classe, sendo seu uso bloqueado externamente. Um elemento pblico pode ser acessado e modificado livremente. Em Python todas as propriedades e todos os mtodos so pblicos, a no ser que voc defina um modificador de acesso. Conhecendo tais conceitos, vamos atualizar a representao de nossa classe.

```
    classe Soma:
        Propriedades:
            Elemento1 (Pblica),
            Elemento2 (Pblica),
            Resultado (Pblica)
        Mtodos:
            Soma(Elemento1, Elemento2) (Pblico), # Construtor
            exibirOperao() (Pblico)
```

Logo de cara percebemos que a propriedade `Resultado` no deveria ser livre, tampouco as propriedades `Elemento1` e `Elemento2` (imagine se alterarmos os elementos de uma soma e mantermos o resultado, na maioria dos casos estaramos "estragando" a soma). Os mtodos, no entanto, devem permanecer pblicos j que  de nosso interesse que o usurio possa acess-los sempre que quiser. Isso  recorrente em muitas classes, ou seja, os mtodos costumam ser pblicos e as propriedades costumam ser privadas. Vamos atualizar nossa classe, agora sem especificar quando o elemento  pblico:

```
    classe Soma:
        Propriedades:
            Elemento1 (Privada),
            Elemento2 (Privada),
            Resultado (Privada)
        Mtodos:
            Soma(Elemento1, Elemento2), # Construtor
            exibirOperao()
```

Mas e se quisermos, por exemplo, apenas realizar um simples `print(soma.Resultado)` ou ento modificar os elementos corretamente? Para esse fim criaremos mtodos conhecidos como _**getters**_ e _**setters**_. _Getters_ so responsveis por retornar os valores das propriedades e _setters_ so responsveis pela modificao dessas propriedades. Definindo nossos _getters_ e _setters_:

```
    classe Soma:
        Propriedades:
            Elemento1 (Privada),
            Elemento2 (Privada),
            Resultado (Privada)
        Mtodos:
            Soma(Elemento1, Elemento2), # Construtor
            modificarSoma(Elemento1, Elemento2), # Setter (Define novos valores para Elemento1 e Elemento2)
            receberElementos(), # Retorna os valores de Elemento1 e Elemento2
            receberResultado() # Retorna o valor de Resultado
```

 importante frisar que dentro do mtodo `modificarSoma` no estamos simplesmente modificando `Elemento1` e `Elemento2`. Estamos, na realidade, fazendo algo do tipo:

```
    mtodo modificarSoma(NovoElemento1, NovoElemento2):
        Elemento1 = NovoElemento1
        Elemento2 = NovoElemento2
        Resultado = NovoElemento1 + NovoElemento2
```

### 2.3. Herana

Ainda em nosso exemplo da classe `Soma`, podemos definir a soma como herdeira da classe `Operao`. Mas como assim herdeira? Veja bem, existem quatro operaes matemticas bsicas (adiao, multiplicao, subtrao e diviso) e todas elas possuem caractersticas em comum. As caractersticas comuns das operaes matemticas so:

* Possuem dois elementos (operandos)
* Possuem um resultado

Portanto, podemos definir uma classe mais genrica chamada `Operao` da seguinte forma:

```
    classe Operao:
        Propriedades:
            Elemento1 (Privada),
            Elemento2 (Privada),
            Resultado (Privada)
        Mtodos:
            Operao(Elemento1, Elemento2),
            modificarElementos(Elemento1, Elemento2),
            receberElementos(),
            receberResultado()
```

Podemos agora recriar a classe `Soma` como **herdeira** da classe `Operao`:

```
    classe Soma (herdeira de Operao):
        Mtodos:
            Soma(Elemento1, Elemento2),
            modificarElementos(Elemento1, Elemento2)
```

Perceba que no declaramos nenhuma propriedade. Com a herana, a classe `Soma` **herda** todos os mtodos e propriedades da classe `Operao`. Com isso precisamos apenas **sobrescrever** os mtodos que tero comportamento diferente (uma soma no funciona da mesma forma que uma diviso e, portanto, seus construtores e _setters_ so diferentes).

### 3.2. Mtodos

Como dito antes, mtodos so as "funes" das classes. Vamos colocar a mo na massa e criar o mtodo `exibir_operacao`:

````python
class Soma:
    """
    Propriedades:
        elemento1 : primeiro elemento da operao (int ou float)
        elemento2 : segundo elemento da operao (int ou float)
        resultado : resultado da operao (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.elemento1 = elemento1
        self.elemento2 = elemento2
        self.resultado = elemento1 + elemento2
    
    def exibe_operacao(self):
        string_operacao = f"{self.elemento1} + {self.elemento2} = {self.resultado}"
        print(string_operacao)
````

````python
soma = Soma(2, 2)
soma.exibe_operacao()
````

### 4.1. Encapsulamento

Retomando nosso exemplo da classe `Soma` presente nas sees 2 e 3, voc j deve ter percebido que nossa classe `Soma` ainda no possui nenhum tipo de encapsulamento e, portanto, podemos alterar seus parmetros livremente. Nossa primeira tarefa  definir os `getters` e `setters`. Note que j definimos um setter (o mtodo `modificar_elementos`), precisamos agora definir os getters.

Uma coisa importante que precisamos definir  como chamaremos esses mtodos, existem duas formas de declarar (e posteriormente chamar) _getters_ e *setters*, observe a tabela abaixo:

<table>
    <tr>
        <th></th>
        <th>Jeito no pythnico</th>
        <th>Jeito pythnico</th>
    </tr>
    <tr>
        <td>GET</td>
        <td>get_parametro()</td>

Primeiro definiremos do jeito no-_pythnico_ ( mais simples, mas no to elegante):

```python
class Soma:
    [...]
    def get_elemento1(self):
        return self.elemento1
        
    def get_elemento2(self):
        return self.elemento2
        
    def get_resultado(self):
        return self.resultado
```

O jeito _pythnico_ de se fazer isso  utilizando um mecanismo da linguagem chamado ___decorator___, que auxilia o interpretador e pode ser muito til em seu cdigo. No se preocupe se no conseguir entender, fazer as coisas do jeito no-_pythnico_ no tornar seu cdigo inutilizvel.

Para definir os _getters_ utilizamos o _decorator_ `@property` uma linha antes de definir o mtodo.

```python
class Soma:
    [...]
    @property
    def elemento1(self):
        return self.elemento1
        
    @property
    def elemento2(self):
        return self.elemento2
        
    @property
    def resultado(self):
        return self.resultado
```

Para os _setters_ utilizamos o _decorator_ `@nome_da_propriedade.setter`:

```python
class Soma:
    [...]
    @elemento1.setter
    def elemento1(self, valor):
        self.elemento1 = valor
        self.resultado = valor + self.elemento2
        
    @elemento2.setter
    def elemento2(self, valor):
        self.elemento2 = valor
        self.resultado = self.elemento2 + valor
```

Desse modo no precisamos mais do mtodo `modificar_elementos`. Vamos ver como ficou nossa classe:


Ainda no podemos usar os recm criados _getters_ e _setters_ (o interpretador apontar erros). Precisamos definir nossas propriedades como privadas. Para isso devemos inserir um _underscore_ duplo (`__`) antes do nome de nossa propriedade. Dessa maneira:

```python
class Soma:
    """
    Propriedades:
        __elemento1 : primeiro elemento da operao (int ou float)

        self.__elemento1 = elemento1
        self.__elemento2 = elemento2
        self.__resultado = elemento1 + elemento2

        return string_operacao

    def elemento1(self):
        return self.__elemento1

    @property
    def elemento2(self):
        return self.__elemento2

    @property
    def resultado(self):
        return self.__resultado
    
    @elemento1.setter
    def elemento1(self, valor):
        self.__elemento1 = valor
        self.__resultado = valor + self.__elemento2

    @elemento2.setter
    def elemento2(self, valor):
        self.__elemento2 = valor
        self.__resultado = self.__elemento1 + valor
```

Vamos colocar a classe para funcionar:

````python
class Soma:
    """
    Propriedades:
        __elemento1 : primeiro elemento da operao (int ou float)
        __elemento2 : segundo elemento da operao (int ou float)
        __resultado : resultado da operao (int ou float)
    """
    def __init__(self, elemento1, elemento2):
        self.__elemento1 = elemento1
        self.__elemento2 = elemento2
        self.__resultado = elemento1 + elemento2

    def __str__(self):
        string_operacao = f"{self.__elemento1} + {self.__elemento2} = {self.__resultado}"
        return string_operacao

    @property
    def elemento1(self):
        return self.__elemento1

    @property
    def elemento2(self):
        return self.__elemento2

    @property
    def resultado(self):
        return self.__resultado

    @elemento1.setter
    def elemento1(self, valor):
        self.__elemento1 = valor
        self.__resultado = valor + self.__elemento2

    @elemento2.setter
    def elemento2(self, valor):
        self.__elemento2 = valor
        self.__resultado = self.__elemento1 + valor
````

### 4.2. Herana

Para uma perspectiva prtica de herana iremos criar a classe `Operacao` que realiza operaes matemticas.
```
classe Operacao:
    Propriedades:
        num1 : (inteiro) primeiro nmero da operao
        oper : (caractere) operador, define a operao a ser realizada
        num2 : (inteiro) segundo nmero da operao
    Mtodos:
        resultado : retorna o resultado da operao
```

````python
class Operacao:
    """
    Propriedades:
        num1 : (int) primeiro nmero da operao
        oper : (str) operador, define a operao a ser realizada
        num2 : (int) segundo nmero da operao
    """
    def __init__(self, num1, oper, num2):
        self.num1 = num1
        self.oper = oper
        self.num2 = num2
        
    def resultado(self):
        if self.oper == "+":
            return self.num1 + self.num2
        elif self.oper == "-":
            return self.num1 - self.num2
        elif self.oper == "*":
            return self.num1 * self.num2
        elif self.oper == "/":
            return self.num1 / self.num2
````

````python
# Clula testadora
oper = Operacao(2, "*", 2) # 2 x 2
oper.resultado() # Deve retornar 4
````

Temos uma classe genrica de operaes matemticas, podemos criar a classe `Soma` a partir dessa classe. Basta herdarmos ela quando implementarmos `Soma`.

````python
class Soma(Operacao): # Definimos Soma como filha (ou herdeira) de Operacao
    # Ao definirmos isso, Soma pode acessar mtodos e propriedades privadas de Operacao
    # Soma tambm herda mtodos e propriedades pblicas
    """
    Propriedades:
        num1 : (int) primeiro nmero da soma
        num2 : (int) segundo nmero da soma
    """
    def __init__(self, num1, num2):
        super().__init__(num1, "+", num2) # Referncia  classe me (aquela da qual Soma herda suas caractersticas)
````

````python
oper = Soma(3, 2) # 3 + 2
oper.resultado() # Deve retornar 5
````

Perceba como  simples, no precisamos definir o mtodo `resultado` em `Soma` e seu funcionamento foi excelente, uma vez que o objeto gerou o mesmo resultado que `Operacao(num1, "+", num2)`, e foi exatamente isso que criamos. A palavra chave `super` referenciou a classe me, e podemos provar isso implementando de outra forma:

````python
class Soma(Operacao):
    """
    Propriedades:
        num1 : (int) primeiro nmero da soma
        num2 : (int) segundo nmero da soma
    """
    def __init__(self, num1, num2):
        Operacao.__init__(self, num1, "+", num2)
````

A classe abaixo representa um tringulo retngulo. Como bem sabemos um tringulo retngulo de catetos com, respectivamente, $a$ e $b$ de comprimento e hipotenusa de $c$ de comprimento deve seguir a frmula:
$c^2 = a^2 + b^2$

````python
import math
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """
    def __init__(self, cateto_a, cateto_b):
        self.cateto_a = cateto_a
        self.cateto_b = cateto_b
        self.hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    @property
    def cateto_a(self):
        return self.cateto_a
    
    @property
    def cateto_b(self):
        return self.cateto_b
    
    @property
    def hipotenusa(self):
        return self.hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.cateto_a = valor
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.cateto_b = valor
        
TrianguloRetangulo(3, 4)
````

Observe que obtemos uma mensagem de erro do tipo `AttributeError`. Ela nos indica que estamos tentando definir o valor de um atributo sem que o seu `setter` tenha sido definido apropriadamente. Isso ocorreu pois estamos atribuindo valor  propriedade declarada com o decorator `@property` e no  propriedade privada, que nem mesmo chegou a ser declarada. Podemos resolver este problema da seguinte forma:

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__hipotenusa = (cateto_a**2 + cateto_b**2)**(1/2)
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    @property
    def cateto_a(self):
        return self.__cateto_a
    
    @property
    def cateto_b(self):
        return self.__cateto_b
    
    @property
    def hipotenusa(self):
        return self.__hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.__cateto_a = valor
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.__cateto_b = valor
````

Perceba que alteramos o valor do cateto A e ainda assim a hipotenusa manteve seu valor original. Isso  um erro comum ao lidar com objetos e entra na categoria dos erros de semntica, voc no ser notificado pelo interpretador e ele ir aparecer na pior hora possvel. Devemos atualizar o valor da hipotenusa sempre que alterarmos o valor de `cateto_a` e `cateto_b`.

````python
class TrianguloRetangulo:
    """
    Propriedades:
        cateto_a = comprimento do cateto A em "unidades de comprimento"
        cateto_b = comprimento do cateto B em "unidades de comprimento"
        hipotenusa = comprimento da hipotenusa em "unidades de comprimento"
    """
    def __init__(self, cateto_a, cateto_b):
        self.__cateto_a = cateto_a
        self.__cateto_b = cateto_b
        self.__atualiza_hipotenusa()
        
    def __str__(self):
        return "a:{}, b:{}, c:{}".format(self.cateto_a, self.cateto_b, self.hipotenusa)
    
    def __atualiza_hipotenusa(self):
        self.__hipotenusa = (self.cateto_a**2 + self.cateto_b**2)**(1/2)
    
    @property
    def cateto_a(self):
        return self.__cateto_a
    
    @property
    def cateto_b(self):
        return self.__cateto_b
    
    @property
    def hipotenusa(self):
        return self.__hipotenusa

    @cateto_a.setter
    def cateto_a(self, valor):
        self.__cateto_a = valor
        self.__atualiza_hipotenusa()
        
    @cateto_b.setter
    def cateto_b(self, valor):
        self.__cateto_b = valor
        self.__atualiza_hipotenusa()
````
