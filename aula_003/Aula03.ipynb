## 1. Os operadores de comparao

No mundo da programao, um conceito muito til e verstil  o da comparao. Em Python temos alguns operadores que conseguem realizar comparaoes entre valores. Eles so:

`a > b    #maior que`<br>
`a < b    #menor que`<br>
`a >= b   #maior ou igual a`<br>
`a <= b   #menor ou igual a`<br>
`a == b   #igual a`<br>
`a != b   #diferente de`<br>

Com esses operadores podemos realizar comparaes com diversos tipos diferentes, comparando valores numricos, strings, etc. Essas comparaes so chamadas de expresses lgicas ou booleanas e so avaliadas como sendo verdadeiras ou falsas. Em Python, essas expresses lgicas retornam um tipo Booleano com valor `True` (verdadeiro) ou `False` (falso).

Vale resaltar que o operador "=" se comporta como um operador de atribuio e no de igualdade como  o caso do "==".

Vamos observar os seguintes exemplos:

```python
#vamos considerar as seguintes variveis
a = 10
b = 20.5

#agora vamos testar os nossos operadores de comparao. Troque os comparadores e veja qual o resultado!

a > b
```

Muito bom! Mas como falamos (e talvez voc j tenha testado) essas comparaes conseguem fazer mais que comparar valores numricos:

```python
#Vamos agora considerar nossas variveis 'a' e 'b' com o tipo string
a = 'aaa'
b = 'aab'

#Vamos testar se nossos operadores ainda funcionam

a < b
```

Conseguiu? Provavelmente no, n? Isso  por que ainda falta uma pea nesse quebra-cabeas de comparao: Algo que consiga avaliar o resultado dessa operao!<br>

Precisamos de uma maneira de pedir para o computador fazer alguma tarefa baseado no resultado da comparao entre os valores. Para isso usamos o termo `if`, que vem do ingls `se`. Com esse termo, podemos pedir para o computador fazer `alguma coisa` `se` uma `condio` for atendida. Vamos tentar entender melhor:<br>

   Precisamos fazer com  que o computador<br>
    `imprima o valor 'a'` `se` `o valor 'a' for maior que o valor 'b'`<br>
     ou<br>
    `imprima o valor 'b'` `se` `o valor 'b' for maior que o valor 'a'`<br>
    
    
Entendeu? Agora vamos ver como escrever coisas desse tipo em um programa de verdade.<br>


Em Python utilizamos a expresso `if` da seguinte forma:<br>
`if 'Alguma condio':
    'Alguma ao'`<br>

Perceba o espaamento que existe da linha em que a ao est descrita em relao ao lado esquerdo. Essa prtica, chamada de indentao,  fundamental para que cdigos em Python funcionem corretamente (sendo optativa em outras linguages de programao). A tecla ideal para fazer este espaamento  o TAB (em cima de Caps Lock), porm tambm pode ser feito com a barra de espao.

A condio pode ser qualquer valor ou expresso lgica. Quando a condio retornar o valor booleano `True` ou qualquer valor diferente de zero, a ao ser executada. Caso o retorno seja o booleano `False`, `None` ou 0 (zero), a ao no  executada e o programa segue normalmente. Vamos testar:

```python
#Vamos considerar a varivel

a = True

#Troque o valor de 'a' e veja os casos em que seu valor  considerado verdadeiro
if a:
    print('A condio  verdadeira')

```

Abaixo temos um exemplo utilizando os operadores de comparao j ensinados:

```python
#Troque o valor destas variveis e verifique as diferentes sadas de acordo
#com as comparaes realizadas
meuDinheiro = 2.50
precoCoxinha = 2

if meuDinheiro > precoCoxinha:
    print("Vou comer uma coxinha e ainda vai sobrar dinheiro!")
    
if meuDinheiro == precoCoxinha:
    print("Ser que eu devo gastar todo meu dinheiro em uma coxinha?")
    
print("Com fome ou no, vida que segue")
```

O mais interessante de se observar neste exemplo  que, quando a expresso lgica do `if`  falsa, o programa continua sendo executado e apenas ignora a ao englobada.

O operador `elif`  uma contrao do termo `else if` (se voc j viu linguagens como C talvez conhea dessa forma) e significa `caso contrrio, se`. Ele  muito til para checar mais de uma condio e criar mais de duas possibilidades de ao.<br>

Vamos analisar um pouco de sua sintaxe:<br>
`if 'Alguma condio':
    'Alguma ao'`<br>
`elif 'Alguma outra condio':
    'Alguma outra ao'`   
`else:
    'Alguma outra ao, caso nenhuma das condies sejam atendidas'`<br>
        
Podemos escrever quantos `elif` quisermos, mas apenas um `if` e um `else` para analisar um conjunto de condies.<br>
        
Entendeu? Ento vamos explorar essa propriedade:<br>

```python
#considere agora as seguintes variveis
fome = 10
sono = 15

#dessa vez, tente criar diferentes condies e mudar os valores de 'fome' e 'sono'. 
#Tente tambm adicionar elifs com diferentes condies mais especficas

if fome < sono:
    print('Vou voltar a dormir')
    
elif fome > sono:
    print('Vou pegar alguma coisa para comer')
    
else:
    print('Algum traz comida pra mim?')
```

Um ltimo tema a ser abordado  o aninhamento de expresses `if`, `else` e `elif`:<br>

Podemos escrever expresses do tipo `if` dentro delas mesmas. Com isso, podemos criar condies dentro de condies, filtrando ainda mais nossos valores. Podemos olhar outro diagrama para ajudar na compreenso:

```python
#vamos levar em conta trs variveis
a = 1
b = 5
c = 0.5

#mais uma vez, mude as comparaes, valores e tente acrescentar casos de 'elif'
if a <= b:
    if a > c:
        print('Primeiro "if" aninhado  verdadeiro')
    else:
        print('Primeiro "if" aninhado  falso')
else:
    print('Primeiro "if"  falso')
```

Podemos tambm aninhar os operadores lgicos para obtermos resultados diferentes nas operaes. Abaixo temos a explicao de como funciona o aninhamento de `and` e `not` (mas no se prendam s tabelas, com o tempo tudo se torna natural!): <br>

Se temos duas entradas verdadeiras no operador `and`, normalmente teramos que:

|   A  |   B  | Sada |
|:----:|:----:|-------|
| True | True | True  |

Porm, se aplicarmos o operador `not` depois de aplicarmos o operador `and`, temos que a sada ficaria invertida. Tabela da expresso `not (A and B)` para as duas entradas verdadeiras:

|   A  |   B  | Sada de not (A and B) |
|:----:|:----:|-------|
| True | True | False  |

Assim, analisando toda a tabela verdade desses operadores juntos, obtemos:

<table>
    <tr><th> <center>Operador AND com NOT: not (A and B) </center> </th></tr>
<tr><td>

|   A   |   B   | Sada de not (A and B) | 
|:-----:|:-----:|:-----:|
|  True | True  |  False |
|  True | False | True |
| False |  True | True |
| False | False | True |

</td></tr></table>

A ideia aqui  que os operadores fazem sua parte do trabalho separadamente, mas com os resultados de outros operadores. Combinar operadores pode ficar muito complicado e, sem cuidado, podemos estar aplicando esses operadores com valores que no tinhamos inteno. Por isso, lembre-se de usar os parenteses para explicitar quais operaes voc est fazendo! Assim como na matemtica, as operaces entre parenteses acontecem primeiro, ou seja, tm prioridade!


Vamos ento praticar o aninhamento de outros operadores:

```python
#vamos considerar trs variveis
a = 10
b = 5
c = 17

#troque as comparaes, adicione 'ands' e 'ors', tente introduzir o termo 'not' e mude os valores de 'a', 'b' e 'c'
#dessa vez, troque tambem os parenteses, tente imaginar o que aconteceria com e resultado e compare com o resultado final

if not (a > b and a < c):
    print('O primeiro "if"  verdadeiro')

if (a < b or a < c) and (c > b):
    print('O segundo "if"  verdadeiro')
```

## 4. O que pode dar errado?

No decorrer da aula voc provavelmente j teve alguns problemas enquanto programava, e para tentar solucionar as dvidas mais frequentes temos essa seo.

Um dos erros mais frequentes no uso das condicionais em Python  a falta da indentao (IndentationError). Observe o cdigo abaixo:

```python
a = True

if a:
print("123")
```

Deu problema, n? Como dito no durante a aula, a indentao (o espao do comando em relao ao lado esquerdo)  fundamental para a interpretao correta do algoritmo. Caso tenha se esquecido, para dar este espaamento basta apertar a tecla TAB.

Os dois pontos aps a condio tambm  fundamental. Se voc esquecer de utiliz-los, o computador vai indicar um erro.

```python
a = True

if a
    print("123")
```

```python
a = True

if a:
    print("Verdade")
else
    print("Mentira")
```

Outro erro bem comum que observamos em edies anteriores era o uso da comparao de maneira irregular:

```python
x = 5
y = 2
z = 3

if x > y < z:
    print("Mas que loucura  essa?")
```

Apesar do cdigo funcionar,  importante entendermos o que ele significa. Caso a inteno seja de obter o resultado booleano de "x  maior que y e x  menor que z?", o cdigo no est correto. Observe o cdigo abaixo e veja a equivalncia utilizando operadores lgicos, alm de uma soluo para o questionamento.

```python
x = 5
y = 2
z = 3

if x > y and y < z:
    print("Essa  a comparao feita pelo exemplo")
    
if x > y and x < z:
    print("Essa  a comparao citada no texto")
    
if z > x > y:
    print("Essa  uma outra alternativa para a comparao citada no texto (pense um pouco)")
```
