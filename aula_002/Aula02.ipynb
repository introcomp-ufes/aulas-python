## 1. O que é um programa?



Um programa é uma sequência de instruções que especificam como o computador pode realizar uma tarefa. Essa tarefa pode ser algo como encontrar as raízes de uma equação de segundo grau,  uma busca num texto para encontrar e substituir uma palavra, etc. 

Diferentemente de um algoritmo, um programa é escrito em alguma linguagem formal, chamada de **linguagem de programação**. 
Da mesma forma como as pessoas se comunicam umas com as outras por meio de alguma linguagem (português, inglês, francês, etc.), precisamos de um jeito para nos comunicarmos com o computador. Para isso, utilizamos alguma linguagem de programação, como: C, C++, Java, PHP, Lua, etc.

Nesse curso, vamos aprender a linguagem de programação chamada **Python**.

Normalmente, as instruções que compõem um programa variam de acordo com a linguagem de programação que você está utilizando, mas é comum em praticamente todos os programas algumas instruções básicas:

- Entrada: Receber informações, seja do teclado, de um arquivo, do mouse, ou de um sensor, por exemplo;
- Saída: Enviar informações, seja no terminal, num arquivo, em um motor ou display;
- Operações Aritméticas : Realizar operações matemáticas (soma, subtração, divisão e etc) geralmente é essencial para a maioria dos programas;
- Condicionais: Checar condições e realizar instruções apropriadas;
- Repetições: Permite que um número de comandos seja repetido.

Acredite, qualquer programa que você fizer, por mais complicado que seja, é feito de instruções que se aproximam dessas 5 listadas acima.

## 4. Variáveis



Variáveis são espaços na memória destinados a armazenar informações que serão utilizadas no futuro pelo computador.

As variáveis podem ter qualquer nome, respeitando algumas regras:

    1. O primeiro caracter precisa ser uma letra (A-Z, a-z) ou um underscore ( _ )
    2. Os demais caracteres podem ser letras, números ou underscore
    
Há algumas convenções sobre como devemos nomear variáveis; em algumas linguagens é utilizado o *camel case* e, em Python, existe a convenção **PEP8**, a qual sugere que todo o nome da variável deve ser escrito em letras minúsculas e cada palavra deve ser separada por um *underscore* (`_`). Alguns exemplos estão presentes na tabela abaixo:

Nome modelo | Variável em camel case | Variável seguindo a PEP8
------------|------------------------|-------------------------
Dinheiro na conta|dinheiroNaConta    |dinheiro_na_conta
Idade mínima|idadeMinima             |idade_minima
CPF do Cliente|cpfCliente            |cpf_cliente

**É comum evitar acentos e outros caracteres especiais ao criar variáveis!**
    
Além disso, existem algumas palavras reservadas na linguagem que não podem ser utilizadas como variáveis, apesar de respeitarem as regras acima. Algumas dessas palavras são: True, False, continue, and, None, etc.

### 7.1. Função `lower`



A função Lower ira transformar todos os caracteres que estão em maiúsculo para minúsculo( A -> a )

```python
frase = "ISSO É UMA STRING"

frase_modificada = frase.lower();

print(frase_modificada)
```

### 7.2. Função `upper`



A função upper ira transformar todos os caracteres que estão em minúsculo para maiúsculo ( a -> A )

```python
frase = "está tudo pequeno"

frase_modificada = frase.upper();

print(frase_modificada)
```

### 7.3. Função `strip`

Essa função "ajeita" nossa string e é bem interessante para receber entradas como _strings_, uma vez que a chance de existirem caracteres especiais a mais diminui bastante.

Vejamos um exemplo sem `strip`.

```python
palavra = " Introcomp  "
palavra == "Introcomp"
```

Agora veja como fica com o uso de `strip`:

```python
palavra = " Introcomp  ".strip()
palavra == "Introcomp"
```

Você já deve ter percebido que `strip` retira os espaços em branco das extremidades da string.

### 7.4. _f-strings_

Além de tratar nossas _strings_ com `upper`, `lower` e `strip`. Existe uma maneira de utilizar uma _string_ como base para uma frase. Esse recurso foi adicionado na versão 3.6 de Python e é extremamente útil.

Veja um exemplo:

```python
nome = input("Qual seu nome? ").strip()
saudacao = f"Bom dia, {nome}!"

print(saudacao)
```

Perceba que o trecho onde existia um `{nome}` foi substituído pelo conteúdo da variável nome. E é exatamente assim que utilizamos _f-strings_. Basta inserir algo do tipo `{minha_variavel}` no meio de uma _string_. Uma outra característica das _f-strings_ é a necessidade da letra `f` antes do primeiro `"`.

## 11. Operadores Lógicos



Em programas, podemos construir expressões cujos resultados são valores lógicos (**verdadeiro** ou **falso**). Essas expressões são chamadas de **expressões lógicas**, e podem ser compostas de **relações lógicas** e outras expressões lógicas.

Alguns exemplos de relações lógicas:
1. `x == 10` 
2. `idade > 18`
3. `(x+y) < 100`

Veja que utilizamos os operadores de comparação de igualdade (**==**), maior que (**>**) e menor que (**<**) para formar as relações lógicas; o resultado de uma relação pode ser verdadeiro ou falso, dependendo dos valores atribuídos às variáveis.

Agora, podemos construir expressões lógicas que combinam relações lógicas, por exemplo: 
1. `x == 10` **e** `idade > 18`
2. `(x+y) < 100`  **ou** `idade > 18`
3. `x == 10` **e** `(x+y) < 100` **ou** `idade >18`

Note que usamos as palavras **e** e **ou** para combinar relações. Em código Python, essas palavras tornam-se **and** e **or** e são chamados de **operadores lógicos**. Importante observar que o uso de parênteses impõe prioridades, como de costume.

Considerando as expressões acima, elas ficariam assim em Python:

1. `(x == 10) and (idade > 18)`
2. `((x+y) < 100) or (idade > 18)`
3. `((x == 10) and ((x+y) < 100)) or idade > 18`

Para saber os resultadores das expressões lógicas, precisamos entender um pouco mais sobre o comportamento dos operadores **and** e **or**, como veremos na próxima aula.

```python
palavra = input("Qual a palavra a ser decifrada? ") # O usuario irá digitar a entrada que ele quer dar para o programa

# Tratando a palavra:
# A função strip ira retirar os espaços do começo e do fim da nossa entrada digitada
# A função lower ira transformar todos os caracteres que estão em maiúsculo para minúsculo( A -> a )
palavra = palavra.strip().lower()



# Exibindo a palavra:
print(f'A palavra é "{palavra}".')
```

## 13. O que pode dar errado?


Erros acontecem com todo programador. Os programas apresentam diferentes tipos de erros, alguns estão relacionados a erros de sintaxe (geralmente acusados pelo interpretador) e outros a erros de semântica. Erros de semântica (ou seja, de lógica) são mais complicados pois dependem totalmente do programador para os encontrar e solucionar. Mas sempre tem uma forma de resolvê-los. Não desanimem e continuem programando!

Aqui estão alguns exemplos de erros comuns cometidos por programadores:

- Escrever algo da linguagem Python errado. Por exemplo, copie o trecho de código a seguir na celula abaixo e veja o que acontece: `pritn("Por que esse print não está funcionando? :c")`;
- Confundir letras maiúsculas e minúsculas nos nomes de variáveis: somaDosNumeros, SOMADOSNUMEROS, somadosnumeros são 3 variáveis diferentes;
- Indentação e espaços em branco. Siga um modelo de tabulação consistente e único no seu script. Se você começou usando dois espaços, ou tabulação de 4 espaços ou seja o que for, mantenha-a. O interpretador, geralmente, não saberá o que fazer se você misturar vários de tipos de tabulação e espaçamento.



```python
pritn("Por que esse print não está funcionando? :c")
```

Existem estratégias para corrigir o código, as quais são especialmente úteis quando existem erros de semântica (o interpretador normalmente nos alerta sobre erros de sintaxe). Tais estratégias são utilizadas para que encontremos as falhas rapidamente, saber onde elas estão quase sempre é a parte mais complicada.
