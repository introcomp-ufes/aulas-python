Um programa  uma sequncia de instrues que especificam como o computador pode realizar uma tarefa. Essa tarefa pode ser algo como encontrar as razes de uma equao de segundo grau,  uma busca num texto para encontrar e substituir uma palavra, etc.

Essa funo "ajeita" nossa string e  bem interessante para receber entradas como _strings_, uma vez que a chance de existirem caracteres especiais a mais diminui bastante.

Voc j deve ter percebido que `strip` retira os espaos em branco das extremidades da string.

### 7.4. _f-strings_

Alm de tratar nossas _strings_ com `upper`, `lower` e `strip`. Existe uma maneira de utilizar uma _string_ como base para uma frase. Esse recurso foi adicionado na verso 3.6 de Python e  extremamente til.

Veja um exemplo:

```python
nome = input("Qual seu nome? ").strip()
saudacao = f"Bom dia, {nome}!"

print(saudacao)
```

Perceba que o trecho onde existia um `{nome}` foi substitudo pelo contedo da varivel nome. E  exatamente assim que utilizamos _f-strings_. Basta inserir algo do tipo `{minha_variavel}` no meio de uma _string_. Uma outra caracterstica das _f-strings_  a necessidade da letra `f` antes do primeiro `"`.

## 11. Operadores Lgicos



Em programas, podemos construir expresses cujos resultados so valores lgicos (**verdadeiro** ou **falso**). Essas expresses so chamadas de **expresses lgicas**, e podem ser compostas de **relaes lgicas** e outras expresses lgicas.

Alguns exemplos de relaes lgicas:
1. `x == 10` 
2. `idade > 18`
3. `(x+y) < 100`

Veja que utilizamos os operadores de comparao de igualdade (**==**), maior que (**>**) e menor que (**<**) para formar as relaes lgicas; o resultado de uma relao pode ser verdadeiro ou falso, dependendo dos valores atribudos s variveis.

Agora, podemos construir expresses lgicas que combinam relaes lgicas, por exemplo: 
1. `x == 10` **e** `idade > 18`
2. `(x+y) < 100`  **ou** `idade > 18`
3. `x == 10` **e** `(x+y) < 100` **ou** `idade >18`

Note que usamos as palavras **e** e **ou** para combinar relaes. Em cdigo Python, essas palavras tornam-se **and** e **or** e so chamados de **operadores lgicos**. Importante observar que o uso de parnteses impe prioridades, como de costume.

Considerando as expresses acima, elas ficariam assim em Python:

1. `(x == 10) and (idade > 18)`
2. `((x+y) < 100) or (idade > 18)`
3. `((x == 10) and ((x+y) < 100)) or idade > 18`

Para saber os resultadores das expresses lgicas, precisamos entender um pouco mais sobre o comportamento dos operadores **and** e **or**, como veremos na prxima aula.

```python
palavra = input("Qual a palavra a ser decifrada? ") # O usuario ir digitar a entrada que ele quer dar para o programa

# Tratando a palavra:
# A funo strip ira retirar os espaos do comeo e do fim da nossa entrada digitada
# A funo lower ira transformar todos os caracteres que esto em maisculo para minsculo( A -> a )
palavra = palavra.strip().lower()



# Exibindo a palavra:
print(f'A palavra  "{palavra}".')
```

Os programas apresentam diferentes tipos de erros, alguns esto relacionados a erros de sintaxe (geralmente acusados pelo interpretador) e outros a erros de semntica. Erros de semntica (ou seja, de lgica) so mais complicados pois dependem totalmente do programador para os encontrar e solucionar. Mas sempre tem uma forma de resolv-los. No desanimem e continuem programando!

Aqui esto alguns exemplos de erros comuns cometidos por programadores:
