## 1. O que  um programa?

Um programa  uma sequncia de instrues que especificam como realizar uma tarefa. Essa tarefa pode ser algo como encontrar as razes de uma equao de segundo grau,  uma busca num texto para encontrar e substituir uma palavra, etc. Normalmente essas instrues variam de acordo com a linguagem de programao que voc est utilizando, mas  comum em praticamente todos os programas algumas instrues bsicas:

- Sada: Enviar informaes, seja no terminal, num arquivo, em um motor ou display;

## 3. Variveis

Como visto na Aula 1, variveis so espaos na memria destinados a armazenar informaes que sero utilizadas no futuro pelo computador.

As variveis podem ter qualquer nome, respeitando algumas regras:
    1. O primeiro caracter precisa ser uma letra (A-Z, a-z) ou um underscore ( _ )
    2. Os demais caracteres podem ser letras, nmeros ou underscore
    
H algumas convenes sobre como devemos nomear variveis, em algumas linguagens  utilizado o *camel case*, em Python existe a conveno **PEP8**, a qual sugere que todo o nome da varivel deve ser escrito em letras minsculas e cada palavra deve ser separada por um *underscore* (`_`). Alguns exemplos esto presentes na tabela abaixo:

Nome modelo | Varivel em camel case | Varivel seguindo a PEP8
------------|------------------------|-------------------------
Dinheiro na conta|dinheiroNaConta    |dinheiro_na_conta
Idade mnima|idadeMinima             |idade_minima
CPF do Cliente|cpfCliente            |cpf_cliente

 comum evitar acentos e outros caracteres especiais ao criar variveis!
    
Alm disso, existem algumas palavras reservadas na linguagem que no podem ser utilizadas como variveis, apesar de respeitarem as regras acima. Algumas dessas palavras so: True, False, continue, and, None, etc.

```python
# SEGUNDA TENTATIVA:

# Recebendo os valores de a, b e c:
c = int(input("Qual o valor de 'a'? "))
b = int(input("Qual o valor de 'b'? "))
a = int(input("Qual o valor de 'c'? "))

# Calculando o delta:
delta = b**2 - 4*a*c

# Calculando x1 e x2
if delta <= 0: # Verificando se existem razes reais (se delta  menor que zero, no h raiz real)
    x1 = (-b - delta)/(2*a)
    x2 = (-b + delta)/(2*a)
    print("x1 = {} & x2 = {}".format(x1, x2))
    
# Caso no exista raiz real:
else:
    print("No h raiz real! :(")
```

Beleza! Conseguimos resolver dois dos erros, como resolveremos o terceiro? Simples, verificaremos o valor de `delta` e o motivo da verificao do seu valor est falhando.

```python
# TERCEIRA TENTATIVA:

# Recebendo os valores de a, b e c:
c = int(input("Qual o valor de 'a'? "))
b = int(input("Qual o valor de 'b'? "))
a = int(input("Qual o valor de 'c'? "))

# Calculando o delta:
delta = b**2 - 4*a*c

# Isso nos ajudar a saber o valor de delta
print("Delta = {}".format(delta))

# Verificando se existem razes reais:
ha_raiz_real = delta <= 0
    
# Isso nos ajudar a saber o resultado do condicional
print(" raiz real? {}".format(ha_raiz_real))

# Calculando x1 e x2:
if ha_raiz_real:
    x1 = (-b - delta)/(2*a)
    x2 = (-b + delta)/(2*a)
    print("x1 = {} & x2 = {}".format(x1, x2))
    
# Caso no exista raiz real:
else:
    print("No h raiz real! :(")
```

OPA! Descobrimos o erro! O programa est dizendo que para `delta` menor ou igual a zero existem razes reais, mas sabemos que existem razes reais se e somente se `delta`  maior ou igual a zero. Corrigimos a condicional (veremos mais sobre estruturas condicionais mais para frente no curso, apenas observe, no se preocupe em entender tal conceito agora) e esse problema foi resolvido.

Entretanto, descobrimos outro problema, `delta` no bate com os valores esperados. Devemos verificar tudo relacionado ao clculo de seu valor. (`a`, `b`, `c` e `delta` esto envolvidos nisso) 

```python
# QUARTA TENTATIVA:

# Recebendo os valores de a, b e c:
c = int(input("Qual o valor de 'a'? "))
b = int(input("Qual o valor de 'b'? "))
a = int(input("Qual o valor de 'c'? "))

# Calculando o delta:
delta = b**2 - 4*a*c

# Isso nos ajudar a saber o valor de a, b, c e delta
print("A={}\nB={}\nC={}\nDelta={}".format(a, b, c, delta))

# Calculando x1 e x2:
if delta >= 0: # Verificando se existem razes reais (se delta  menor que zero, no h raiz real)
    x1 = (-b - delta)/(2*a)
    x2 = (-b + delta)/(2*a)
    print("x1 = {} & x2 = {}".format(x1, x2))
    
# Caso no exista raiz real:
else:
    print("No h raiz real! :(")
```
