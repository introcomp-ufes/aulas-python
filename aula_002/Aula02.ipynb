## 1. O que é um programa?

Um programa é uma sequência de instruções que especificam como realizar uma tarefa. Essa tarefa pode ser algo como encontrar as raízes de uma equação de segundo grau,  uma busca num texto para encontrar e substituir uma palavra, etc. Normalmente essas instruções variam de acordo com a linguagem de programação que você está utilizando, mas é comum em praticamente todos os programas algumas instruções básicas:

- Entrada: Receber informações, seja do teclado, de um arquivo, do mouse, ou de um sensor, por exemplo;
- Saída: Enviar informações, seja no terminal, num arquivo, em um motor ou display;
- Matemática: Realizar operações matemáticas normalmente é essencial para a maioria dos programas;
- Condicionais: Checar condições e realizar instruções apropriadas;
- Repetições: Realizar instruções repetidamente.

Acredite, qualquer programa que você fizer, por mais complicado que seja, é feito de instruções que se aproximam dessas 5 listadas acima.

## 3. Variáveis

Como visto na Aula 1, variáveis são espaços na memória destinados a armazenar informações que serão utilizadas no futuro pelo computador.

As variáveis podem ter qualquer nome, respeitando algumas regras:
    1. O primeiro caracter precisa ser uma letra (A-Z, a-z) ou um underscore ( _ )
    2. Os demais caracteres podem ser letras, números ou underscore
    
Há algumas convenções sobre como devemos nomear variáveis, em algumas linguagens é utilizado o *camel case*, em Python existe a convenção **PEP8**, a qual sugere que todo o nome da variável deve ser escrito em letras minúsculas e cada palavra deve ser separada por um *underscore* (`_`). Alguns exemplos estão presentes na tabela abaixo:

Nome modelo | Variável em camel case | Variável seguindo a PEP8
------------|------------------------|-------------------------
Dinheiro na conta|dinheiroNaConta    |dinheiro_na_conta
Idade mínima|idadeMinima             |idade_minima
CPF do Cliente|cpfCliente            |cpf_cliente

É comum evitar acentos e outros caracteres especiais ao criar variáveis!
    
Além disso, existem algumas palavras reservadas na linguagem que não podem ser utilizadas como variáveis, apesar de respeitarem as regras acima. Algumas dessas palavras são: True, False, continue, and, None, etc.

### 6.1 A biblioteca 'math'
Já sabemos fazer matemática básica em Python!<br>
Mas e se precisarmos fazer contas matemáticas mais complexas, como potenciação, radiciação, valores absolutos, senos e cossenos, etc?<br>Com sorte, Python tem uma biblioteca própria para isso, a biblioteca "math". Basta digitar `from math import *` no seu código, normalmente no começo do arquivo de código `.py` ou, se você estiver rodando o código em tempo de execução, digite antes de usar algo da biblioteca math pela primeira vez.
Faremos uma explicação mais detalhada sobre bibliotecas e funções no futuro.

Ao utilizar a biblioteca math, devemos nos atentar a alguns detalhes, como a sintaxe das funções matemáticas:<br>
para utilizar uma função, escrevemos: nome_da_funcao(valores)<br><br>
A seguir, alguns exemplos de funções da math:<br><br>
`ceil(valor)` - arredonda o número passado para cima<br><br>
`floor(valor)` - arredonda o número passado para baixo<br><br>
`abs(valor)` - retorna o módulo (valor absoluto) do número inteiro<br><br>
`factorial(valor)` - retorna o fatorial do número<br><br>
`pow(valor1, valor2)` - retorna o valor1 elevado ao valor2<br><br>
`sqrt(valor)` - retorna a raiz quadrada do valor<br><br>
`sin(valor), cos(valor), tan(valor)` - retorna, respectivamente, o seno, o cosseno e a tangente de um valor.<br>NOTA: o valor deve estar em radianos!<br><br>

Além disso, também temos as variáveis "pi" e "e", que também vem com a biblioteca. Por serem variáveis, você pode alterar seus valores do jeito que quiser, mas não é recomendado, uma vez que você pode voltar a precisar desses valores no futuro. Caso você tenha alterado essas variáveis e precise dos valores de novo, basta escrever: <br>
`pi = math.pi`<br>
`e = math.e`<br>

Caso você esteja curioso para ver o resto da documentação da biblioteca math, clique <a href = "https://docs.python.org/3/library/math.html">aqui</a>

```python
import math
```

```python
# Isso é uma potenciação
print(pow(5, 3))
#se você preferir, pode fazer potenciações sem importar a math desse jeito:
print(5**3)
```

```python
# Isso é uma radiciação
print(sqrt(121))
#se você preferir, pode fazer radiciações sem importar a math desse jeito:
print(121**0.5)
```

```python
# Executa o fatorial de um número
print(factorial(5))
```

```python
# Retorna o cosseno de um número (em rad) 
print(cos(0))
```

```python
# Arredonda o número para cima
print(ceil(4.25))
```

```python
# Faz e^4
print(pow(e, 4))
```

## 7. Strings

Além de operações matemáticas, é muito comum a manipulação de strings (textos) em programas. A seguir, algumas técnicas para tratamento de strings em Python:

```python
# Note que tanto aspas simples ('string') e aspas duplas ("string") fazem a mesma coisa!
frase = "isso é uma string"
```

```python
print(frase)
```

```python
# Usando o operador +, podemos concatenar duas strings! (juntar o fim da primeira no início da segunda, formando uma só)
frase = frase + ' em Python'
```

## 11. O que pode dar errado?

Erros acontecem com todo programador. Os programas apresentam diferentes tipos de erros, alguns estão relacionados a erros de sintaxe (geralmente acusados pelo interpretador) e outros a erros de semântica. Erros de semântica (ou seja, de lógica) são mais complicados pois dependem totalmente do programador para os encontrar e solucionar. Mas sempre tem uma forma de resolvê-los. Não desanimem e continuem programando!

Aqui estão alguns exemplos de erros comuns cometidos por programadores:

- Escrever algo da linguagem Python errado. Por exemplo, copie o trecho de código a seguir na celula abaixo e veja o que acontece: `pritn("Por que esse print não está funcionando? :c")`;
<br>

- Confundir letras maiúsculas e minúsculas nos nomes de variáveis: somaDosNumeros, SOMADOSNUMEROS, somadosnumeros são 3 variáveis diferentes;
<br>

- Indentação e espaços em branco. Siga um modelo de tabulação consistente e único no seu script. Se você começou usando dois espaços, ou tabulação de 4 espaços ou seja o que for, mantenha-a. O interpretador, geralmente, não saberá o que fazer se você misturar vários de tipos de tabulação e espaçamento.



```python
pritn("Por que esse print não está funcionando? :c")
```

Existem estratégias para corrigir o código, as quais são especialmente úteis quando existem erros de semântica (o interpretador normalmente nos alerta sobre erros de sintaxe). Tais estratégias são utilizadas para que encontremos as falhas rapidamente, saber onde elas estão quase sempre é a parte mais complicada.

#### Exemplo:
Façamos um programa para calcular as raízes de uma equação de segundo grau da forma `ax² + bx + x = 0`. As entradas e saídas do nosso programa devem ser as mesmas da tabela abaixo, ou seja, para cada valor de `a`, `b` e `c` na entrada a saída do seu programa deve ser a mesma da tabela. Tarefa simples, não é mesmo?

a | b | c | Saída Esperada
--|---|---|---------------
1 | 2 | 3 | Não há raiz real! :(
3 | 2 | 1 | Não há raiz real! :(
-1 | -2 | -3 | Não há raiz real! :(
1 | 0 | 0 | x1 = 0.0 & x2 = 0.0
1 | 5 | 3 | x1 = 4.0 & x2 = -9.0
4 | 9 | 3 | x1 = 3.0 & x2 = -5.25

**NOTA:** A explicação abaixo contém conceitos ainda não explicados no curso, eles estão presentes apenas para fazer com que o exemplo funcione, sem eles teríamos erros.

```python
# PRIMEIRA TENTATIVA:

# Recebendo os valores de a, b e c:
c = int(input("Qual o valor de 'a'? "))
b = int(input("Qual o valor de 'b'? "))
a = int(input("Qual o valor de 'c'? "))

# Calculando o delta:
delta = b**2 - 4*a*c

# Calculando x1 e x2:
if delta <= 0: # Verificando se existem raízes reais (se delta é menor que zero, não há raiz real)
    x1 = (-b - delta)/(2*a)
    x2 = (-b + delta)/(2*a)
    print("x1={} & x2={}".format(x1, x2))
    
# Caso não exista raiz real:
else:
    print("não existe raiz real")
```

Teste essa resolução para as entradas dadas na tabela e perceba que os resultados diferem. Há algo de errado, não é mesmo?

Vejamos as diferenças entre as saídas:
- O programa diz não haver raiz real quando há e diz haver raiz real quando não há, as únicas exceções acontecem quando `delta == 0` e quando `b == 0`, nesse último caso ocorre erro de divisão por zero.
- A mensagem que deveria ser `Não há raiz real! :(` é, na verdade, `não existe raiz real`.
- Perceba que na saída esperada há um espaçamento entre o `=` e os outros caracteres, já na saída obtida isso não existe.

Começaremos a resolver pelo mais fácil, onde já sabemos que está o erro, nos `print()`. Corrigindo tais problemas obtemos o código abaixo:

```python
# SEGUNDA TENTATIVA:

# Recebendo os valores de a, b e c:
c = int(input("Qual o valor de 'a'? "))
b = int(input("Qual o valor de 'b'? "))
a = int(input("Qual o valor de 'c'? "))

# Calculando o delta:
delta = b**2 - 4*a*c

# Calculando x1 e x2
if delta <= 0: # Verificando se existem raízes reais (se delta é menor que zero, não há raiz real)
    x1 = (-b - delta)/(2*a)
    x2 = (-b + delta)/(2*a)
    print("x1 = {} & x2 = {}".format(x1, x2))
    
# Caso não exista raiz real:
else:
    print("Não há raiz real! :(")
```

Beleza! Conseguimos resolver dois dos erros, como resolveremos o terceiro? Simples, verificaremos o valor de `delta` e o motivo da verificação do seu valor está falhando.

```python
# TERCEIRA TENTATIVA:

# Recebendo os valores de a, b e c:
c = int(input("Qual o valor de 'a'? "))
b = int(input("Qual o valor de 'b'? "))
a = int(input("Qual o valor de 'c'? "))

# Calculando o delta:
delta = b**2 - 4*a*c

# Isso nos ajudará a saber o valor de delta
print("Delta = {}".format(delta))

# Verificando se existem raízes reais:
ha_raiz_real = delta <= 0
    
# Isso nos ajudará a saber o resultado do condicional
print("É raiz real? {}".format(ha_raiz_real))

# Calculando x1 e x2:
if ha_raiz_real:
    x1 = (-b - delta)/(2*a)
    x2 = (-b + delta)/(2*a)
    print("x1 = {} & x2 = {}".format(x1, x2))
    
# Caso não exista raiz real:
else:
    print("Não há raiz real! :(")
```

OPA! Descobrimos o erro! O programa está dizendo que para `delta` menor ou igual a zero existem raízes reais, mas sabemos que existem raízes reais se e somente se `delta` é maior ou igual a zero. Corrigimos a condicional (veremos mais sobre estruturas condicionais mais para frente no curso, apenas observe, não se preocupe em entender tal conceito agora) e esse problema foi resolvido.

Entretanto, descobrimos outro problema, `delta` não bate com os valores esperados. Devemos verificar tudo relacionado ao cálculo de seu valor. (`a`, `b`, `c` e `delta` estão envolvidos nisso) 

```python
# QUARTA TENTATIVA:

# Recebendo os valores de a, b e c:
c = int(input("Qual o valor de 'a'? "))
b = int(input("Qual o valor de 'b'? "))
a = int(input("Qual o valor de 'c'? "))

# Calculando o delta:
delta = b**2 - 4*a*c

# Isso nos ajudará a saber o valor de a, b, c e delta
print("A={}\nB={}\nC={}\nDelta={}".format(a, b, c, delta))

# Calculando x1 e x2:
if delta >= 0: # Verificando se existem raízes reais (se delta é menor que zero, não há raiz real)
    x1 = (-b - delta)/(2*a)
    x2 = (-b + delta)/(2*a)
    print("x1 = {} & x2 = {}".format(x1, x2))
    
# Caso não exista raiz real:
else:
    print("Não há raiz real! :(")
```

Perceba que `a`, `b` e `c` estão invertidos, basta corrigir isso no código. Contudo, ainda existem erros, os valores de `x1` e `x2` estão invertidos, basta resolver mais esse problema e notar que tudo funciona conforme esperado.

```python
# PROGRAMA SEM ERROS

# Recebendo os valores de a, b e c:
a = int(input("Qual o valor de 'a'? "))
b = int(input("Qual o valor de 'b'? "))
c = int(input("Qual o valor de 'c'? "))

# Calculando o delta:
delta = b**2 - 4*a*c

# Calculando x1 e x2:
if delta >= 0: # Verificando se existem raízes reais (se delta é menor que zero, não há raiz real)
    x1 = (-b + delta)/(2*a)
    x2 = (-b - delta)/(2*a)
    print("x1 = {} & x2 = {}".format(x1, x2))
    
# Caso não exista raiz real:
else: # Estudaremos em aulas futuras
    print("Não há raiz real! :(")
```
