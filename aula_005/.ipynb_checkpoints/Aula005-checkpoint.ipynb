### Problema 12:
Leonardo Fibonacci foi um grande matemtico italiano. Fibonacci desenvolveu uma forma de descrever o crescimento $F$ de uma populao de coelhos em 1202. O crescimento em um momento $n$  denotado por $F_n$.


<center><b>Figura 7:</b> Um coelho (Imagem de Heidelbergerin por Pixabay)</center>

Num primeiro momento (quando $n = 0$, ou seja $F_0$), no h crescimento nenhum, ou seja $F_0 = 0$. Logo em seguida comeam a surgir os primeiros coelhinhos e, a partir disso, a populao de coelhos aumenta cada vez mais, uma vez que cada casal de coelhos produz uma quantidade de filhotes num determinado tempo.

Fibonacci chegou  seguinte sequncia: $(0, 1, 1, 2, 3, 4, 5, 8, 13, ..., {\infty})$. Podemos descrever tal sequncia como:

$

\text{Se }{n = 0}, F_n = 0 \\
\text{Se }{n = 1}, F_n = 1 \\
\text{Se }{n \geq 2}, F_n = F_{n - 1} + F_{n - 2}
\end{cases}
$

Considere $F$ como uma lista de $N$ elementos. Escreva um programa no qual sua entrada seja constituda de $N \geq 1$ e a sada deve ser a sequncia de Fibonacci em $N$ linhas, com cada linha contendo um elemento da lista.

**Entrada:**\

**Sada:**\
