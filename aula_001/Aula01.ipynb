### 1.2 Expresses aritmticas

A expresso `( 2 + 2 ) * 4`  equivalente  `16` <br>

### 2.2 Estruturas condicionais

Elas possibilitam a escolha de um grupo de aes e estruturas a serem executadas quando determinadas condies so ou deixam de ser satisfeitas. **Por exemplo**: 

-  Se a via estiver livre, acelere, caso contrrio permanea parado.

-  As estruturas condicionais acontecem quando a condio  verdadeira
-  Uma condio s pode ser **verdadeira** ou **falsa**

ou


<center><b>Figura 4:</b> Representao de estrutura condicional.</center>

ou, em pseudocdigo:

`se ( condio ) ento:
    <instrues>`
    <br>
`seno:
    <instrues>`

Observe como se a condio for verdadeira, o caminho que voc ir seguir ser o da esquerda. Se a condio for falsa, ou seja, ela no ocorre, o caminho que voc ir seguir  outro.

### 2.3 Estruturas de repetio

Elas possibilitam que um grupo de aes e estruturas sejam executadas enquanto (ou at que) determinadas condies deixam de ser satisfeitas. **Por exemplo**:


<center><b>Figura 5:</b> Representao de estrutura de repetio.</center>

#### Exemplo 4:

Esse algoritmo nos permite apresentar (imprimir na tela) todos os nmeros de 1 at 20:

`numero = 1`
<br>
`enquanto ( numero <= 20 ) faa:
    exiba numero
    numero = numero + 1`
    
Neste exemplo comeamos a contar com o valor de numero em 1. Como queremos contar at 20, temos que mostrar os nmeros de 1 at 20.<br><br>
Partindo do nmero 1, fazemos uma pergunta: o valor de numero  menor que ou igual a 20? **Sim**. Ento, exibimos o valor de numero (que  1), e guardamos o prximo valor, que ser o valor da varivel numero mais 1, ou seja, 2. Fizemos nossas instrues e partimos novamente para nossa condio e repetimos nossa pergunta: o valor da varivel numero  menor que ou igual a 20? A resposta continua sim! Ento entramos novamente no nosso conjunto de instrues, exibimos o valor da varivel e acrescentamos 1.<br><br>
Continuamos fazendo isto at que o valor de numero seja igual a 20. Quando ele entrar novamente, ele ir exibir o valor da varivel numero (que ser 20) e adicionar 1, tornando numero igual a 21. Quando for fazer a pergunta: o valor da varivel numero  menor que ou igual a 20? A resposta ser **no** e vamos parar de executar essas instrues. Contamos de 1 at 20.
