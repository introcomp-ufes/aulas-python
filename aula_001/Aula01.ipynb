# 1. Algoritmos




Ao cozinhar, você é capaz de seguir as instruções de uma receita? Então você faz o que seu computador faz o tempo todo. Você seguiu um algoritmo!

Um **algoritmo** é a descrição de uma **sequência de passos** finitos (ações/instruções/ordens) que devem ser seguidos em uma determinada ordem para chegar à **solução de um problema**.

## Motivação: por que estudar algoritmos?

Eles nos permitem colocar os computadores encarregados de tarefas repetitivas e em grande escala, que seriam inviáveis de serem feitas por humanos. Também habilitam motores de busca, estudos de química, biologia, simulações, jogos e uma infinidade de outras aplicações.

## 1.1 Entrada e saída


<a id='section1.1'></a>
Tudo o que os computadores fazem é executar diversas operações baseadas em certas informações fornecidas a ele. Essas informações são as **entradas**. Elas permitem ao usuário a comunicação entre o mundo exterior e a máquina executando o programa. Informações fornecidas pelo usuário são, na maioria das vezes, armazenadas no que chamamos de **variáveis**. Por outro lado, as **saídas** são os resultados esperados do programa, podendo ser mensagens, expressões e valores armazenados em variáveis.  Por exemplo, em uma função 
y = x + 3, **x** é o valor de entrada, enquanto **y**, valor que desejo descobrir, é a saída.

Considerando o **Exemplo 1**, informações de mapa, as posições inicial e atual, o destino e o meio de deslocamento formam as entradas para o algoritmo. Enquanto a saída é o resultado da aplicação de seu algoritmo, no caso, qual seria o melhor caminho a ser realizado pelo usuário para chegar em seu destino.

**Exemplo 2:**

Descobrir a hipotenusa de um triângulo retângulo dado o tamanho de dois catetos.


[](attachment:aula1.png)
<center></center>

<center><b>Figura 2:</b> Triângulo retângulo com seus respectivos lados nomeados.</center>

 Algoritmo
1. Eleve o primeiro cateto ao quadrado;
2. Eleve o segundo cateto ao quadrado;
3. Some o resultado do 1° passo com o resultado do 2° passo;
4. Tire a raiz quadrada do resultado do 3° passo;
5. O resultado do 4° passo é a hipotenusa.

É fundamental que saibamos as entradas e as saídas dos problemas que queremos resolver. Neste algoritmo, nosso objetivo é encontrar a hipotenusa (que configura como nossa saída) utilizando obrigatoriamente de duas entradas: a medida do cateto 1 e do cateto 2.

Assim como um problema de matemática é resolvido com um algoritmo, o programador deve reconhecer as entradas do seu programa para elaborar o algoritmo correto que o leva às saídas esperadas.

## 1.2 Expressões aritméticas


<a id='section1.2'></a>

Conforme observamos, para executar um algoritmo é essencial a manipulação de dados, valores ou objetos. Por exemplo, no problema do Mercado, um dado é a posição do carro. No segundo exemplo, o valor de um cateto também é um dado.

Para operar os valores e dados podemos utilizar expressões aritméticas:

-  A operação de soma será representada por `+`;
-  A operação de subtração será representada por `-`;
-  A operação de multiplicação será representada por `*`;
-  A operação de divisão será representada por `/`.
-  A operação de potenciação será representada por `**`.

A ordem de prioridade das operações seguem como o usual:

-  Primeiro as radiciações e potenciações (`**`);
-  Depois, produtos e divisões (`*` e `/`);
-  E, por último, somas e subtrações (`+` e `-`).

Porém, a ordem pode ser alterada pela utilização de parênteses. Então, quando a expressão contiver parênteses, deve ser resolvida do mais interno ao mais externo. Por exemplo:

A expressão `( 2 + 2 ) * 4` é equivalente à `16` <br>
A expressão   `2 + 2 * 4` é equivalente à `10` <br>

## 1.3 Variáveis


<a id='section1.3'></a>


Agora iremos falar sobre a atribuição de tais expressões às variáveis:

-  Quando um computador executa um algoritmo, o mesmo precisa armazenar e manipular múltiplos valores; 
-  Esses valores são armazenados e acessados na memória do computador;
-  As variáveis podem ser entendidas como "caixas" que podem guardar **um** valor;
-  Cada vez que guardamos um valor novo na variável, apagamos o valor antigo;
-  Para guardarmos um valor em uma variável usamos algum **comando de atribuição**;
-  Como uma demonstração, iremos utilizar o sinal de igual **`=`** para definir atribuição de valores;
-  Será estabelecido que a variável a ser atribuída irá se encontrar à esquerda do símbolo;
-  A parte à direita será uma expressão;
-  O resultado da expressão será calculado e posteriormente armazenado na variável.

Vamos ver o exemplo do algoritmo que descobre a hipotenusa de um triângulo retângulo de novo? Dessa vez, vamos utilizar a linguagem Python para escrever o algoritmo. Não se preocupe com a escrita, mas atente-se aos conceitos que aprendeu até agora e como a linguagem os reproduz.

````python
cateto1 = 3
cateto2 = 4

#Passo 1 - Elevamos o primeiro cateto ao quadrado
cateto1 = cateto1**2

#Passo 2 - Elevamos o segundo cateto ao quadrado
cateto2 = cateto2**2

#Passo 3 - Somamos o quadrado dos catetos
soma_do_quadrado_dos_catetos = cateto1 + cateto2

#Passo 4 - Tiramos a raiz quadrada da soma (Um número elevado a 0.5 equivale à sua raíz quadrada)
hipotenusa = soma_do_quadrado_dos_catetos**0.5

#Passo 5 - Temos o resultado utilizando o Teorema de Pitágoras
print(hipotenusa)
````

Além dos códigos das linguagens de programação, podemos representar algoritmos por meio de <b>pseudocódigo.</b> O pseudocódigo nada mais é do que uma maneira "informal" de se escrever código, sem ter uma sintaxe definida e com termos estabelecidos à gosto do autor. Ao longo da aula, diversos exemplos aparecerão em pseudocódigo, para que o entendimento, ao menos nesse início, seja mais claro.

## 2.2 Estruturas condicionais

As estruturas condicionais possibilitam a escolha de um grupo de ações e estruturas a serem executadas quando determinadas condições são ou deixam de ser satisfeitas. *Por exemplo*: 

-  Se a via estiver livre, acelere, caso contrário permaneça parado.

As estruturas condicionais acontecem quando a condição é verdadeira e uma condição só pode ser *verdadeira* ou *falsa*.

[](attachment:estrutura_condicional2.png)
<center></center>

<center><b>Figura 4:</b> Representação de estrutura condicional.</center>

ou, em pseudocódigo:

`se ( condição ) então:
    <instruções>`
    <br>
`senão:
    <instruções>`

Observe como se a condição for verdadeira, o caminho que você irá seguir será o da esquerda. Se a condição for falsa, ou seja, ela não ocorre, o caminho que você irá seguir é outro.

# 3. Linguagens de programação

Para todos os tipos de comunicação são necessárias linguagens. Os computadores apenas são capazes de entender código de máquina, que é de difícil interpretação pelos seres humanos. Porém, ao longo do tempo, foram sendo desenvolvidas linguagens de programação que se aproximam mais da linguagem humana e são "traduzidas" para código de máquinas. Algumas dessas linguagens de programação são: C, C++, Java, Python, PHP, Lua, Ruby, R, Rust, Julia.

A linguagem utilizada nesse curso será o Python.

Essas linguagens que são relativamente mais fáceis de compreender e mais próximas da forma como raciocinamos são classificadas como linguagens de alto nível.

Dentro das linguagens de alto nível, podemos ter linguagens que são interpretadas e outras que são compiladas. Um código que é compilado necessita ser todo traduzido para código de máquina antes de sua execução, enquanto as interpretadas são convertidas linha por linha.
