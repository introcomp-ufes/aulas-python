## 1.1 Entrada e sada


<a id='section1.1'></a>
Tudo o que os computadores fazem  executar diversas operaes baseadas em certas informaes fornecidas a ele. Essas informaes so as **entradas**. Elas permitem ao usurio a comunicao entre o mundo exterior e a mquina executando o programa. Informaes fornecidas pelo usurio so, na maioria das vezes, armazenadas no que chamamos de **variveis**. Por outro lado, as **sadas** so os resultados esperados do programa, podendo ser mensagens, expresses e valores armazenados em variveis.  Por exemplo, em uma funo 
y = x + 3, **x**  o valor de entrada, enquanto **y**, valor que desejo descobrir,  a sada.

## 1.2 Expresses aritmticas


<a id='section1.2'></a>

Conforme observamos, para executar um algoritmo  essencial a manipulao de dados, valores ou objetos. Por exemplo, no problema do Mercado, um dado  a posio do carro. No segundo exemplo, o valor de um cateto tambm  um dado.

Para operar os valores e dados podemos utilizar expresses aritmticas:

-  A operao de soma ser representada por `+`;
-  A operao de subtrao ser representada por `-`;
-  A operao de multiplicao ser representada por `*`;
-  A operao de diviso ser representada por `/`.
-  A operao de potenciao ser representada por `**`.

A ordem de prioridade das operaes seguem como o usual:

-  Primeiro as radiciaes e potenciaes (`**`);
-  Depois, produtos e divises (`*` e `/`);
-  E, por ltimo, somas e subtraes (`+` e `-`).

Porm, a ordem pode ser alterada pela utilizao de parnteses. Ento, quando a expresso contiver parnteses, deve ser resolvida do mais interno ao mais externo. Por exemplo:

A expresso `( 2 + 2 ) * 4`  equivalente  `16` <br>
A expresso   `2 + 2 * 4`  equivalente  `10` <br>

Vamos ver o exemplo do algoritmo que descobre a hipotenusa de um tringulo retngulo de novo? Dessa vez, vamos utilizar a linguagem Python para escrever o algoritmo. No se preocupe com a escrita, mas atente-se aos conceitos que aprendeu at agora e como a linguagem os reproduz.

````python
cateto1 = 3
cateto2 = 4

#Passo 1 - Elevamos o primeiro cateto ao quadrado
cateto1 = cateto1**2

#Passo 2 - Elevamos o segundo cateto ao quadrado
cateto2 = cateto2**2

#Passo 3 - Somamos o quadrado dos catetos
soma_do_quadrado_dos_catetos = cateto1 + cateto2

#Passo 4 - Tiramos a raiz quadrada da soma (Um nmero elevado a 0.5 equivale  sua raz quadrada)
hipotenusa = soma_do_quadrado_dos_catetos**0.5

#Passo 5 - Temos o resultado utilizando o Teorema de Pitgoras
print(hipotenusa)
````

Alm dos cdigos das linguagens de programao, podemos representar algoritmos por meio de <b>pseudocdigo.</b> O pseudocdigo nada mais  do que uma maneira "informal" de se escrever cdigo, sem ter uma sintaxe definida e com termos estabelecidos  gosto do autor.
