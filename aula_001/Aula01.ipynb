## 1. Algoritmos
Você é capaz de seguir essas instruções? Então você fez o que seu computador faz o tempo todo. Você seguiu um algoritmo!

Um **algoritmo** é a descrição de uma **sequência de passos** (ações/instruções/ordens) finitos que devem ser seguidas em sequência para a **solução de um problema**.

#### Motivação: por que estudar algoritmos?

Eles nos permitem colocar os computadores encarregados de tarefas repetitivas e em grande escala, que seriam inviáveis de serem feitas por humanos. Também habilitam motores de busca, estudos de química, biologia, simulações, jogos e uma infinidade de outras aplicações.

<a id='section1.1'></a>
### 1.1 Entrada e saída
Tudo o que os computadores fazem é executar diversas operações baseadas em certas informações fornecidas a ele. Essas informações são as **entradas**. No problema anterior, algumas das entradas são o mapa, a sua posição inicial, sua posição atual, a localização de destino e qual o seu meio de locomoção.
A **saída** do seu algoritmo é o resultado dado pela aplicação do seu algoritmo, ou seja, chegar no Mercado da Madalena.

#### Exemplo 2:

Descobrir a hipotenusa de um triângulo retângulo dado o tamanho de dois catetos.


<center><b>Figura 2:</b> Triângulo retângulo com seus respectivos lados nomeados.</center>

#### Algoritmo
1. Eleve o primeiro cateto ao quadrado;
2. Eleve o segundo cateto ao quadrado;
3. Some o resultado do 1° passo com o resultado do 2° passo;
4. Tire a raiz quadrada do resultado do 3° passo;
5. O resultado do 4° passo é a hipotenusa.

É fundamental que saibamos as entradas e as saídas dos problemas que queremos resolver. Neste algoritmo, nosso objetivo é encontrar a hipotenusa (que configura como nossa saída) utilizando obrigatoriamente de duas entradas: a medida do cateto 1 e do cateto 2.

Assim como um problema de matemática é resolvido com um algoritmo, o programador deve reconhecer as entradas do seu programa para elaborar o algoritmo correto que o leva às saídas esperadas.

### 1.2 Expressões aritméticas

Conforme observamos, para executar um algoritmo é essencial a manipulação de dados, valores ou objetos. Por exemplo, no problema do Mercado, um dado é a posição do carro. No segundo exemplo o valor de um cateto também é um dado.

Para operar os valores e dados podemos utilizar expressões aritméticas:

-  A operação de soma será representada por `+`;
-  A operação de subtração será representada por `-`;
-  A operação de multiplicação será representada por `*`;
-  A operação de divisão será representada por `/`.
-  A operação de potenciação será representada por `**`.

A ordem de prioridade das operações seguem como o usual:

-  Primeiro as radiciações e potenciações (`**`);
-  Depois, produtos e divisões (`*` e `/`);
-  E, por último, somas e subtrações (`+` e `-`).

Porém, a ordem pode ser alterada pela utilização de parênteses. Então, quando a expressão contiver parênteses, deve ser resolvida do mais interno ao mais externo. Por exemplo:

A expressão `( 2 + 2 ) * 4` é equivalente à `16` <br>
A expressão   `2 + 2 * 4` é equivalente à `10` <br>

Vamos ver o exemplo do algoritmo que descobre a hipotenusa de um triângulo retângulo de novo? Dessa vez, vamos utilizar a linguagem Python para escrever o algoritmo. Não se preocupe com a escrita, mas se atente aos conceitos que aprendeu até agora e como a linguagem os reproduz.

```python
cateto1 = 3
cateto2 = 4

#Passo 1 - Elevamos o primeiro cateto ao quadrado
cateto1 = cateto1**2

#Passo 2 - Elevamos o segundo cateto ao quadrado
cateto2 = cateto2**2

#Passo 3 - Somamos o quadrado dos catetos
soma_do_quadrado_dos_catetos = cateto1 + cateto2

#Passo 4 - Tiramos a raiz quadrada da soma (Um número elevado a 0.5 equivale à sua raíz quadrada)
hipotenusa = soma_do_quadrado_dos_catetos**0.5

#Passo 5 - Temos o resultado utilizando o Teorema de Pitágoras
print(hipotenusa)
```

Além dos códigos das linguagens de programação, podemos representar algoritmos por meio de <b>pseudocódigo.</b> O pseudocódigo nada mais é do que uma maneira "informal" de se escrever código, sem ter uma sintaxe definida e com termos estabelecidos à gosto do autor. ## 2. Estruturas de fluxo

Um algoritmo é constituído por estruturas de fluxo, essas estruturas apresentam um conjunto de ações, informações ou instruções e se comportam de diferentes maneiras, sendo 3 delas:

-  Estruturas sequenciais
-  Estruturas condicionais
-  Estruturas de repetição

<a id='section2.1'></a>

### 2.1 Estruturas sequenciais

Estrutura sequencial é um conjunto de instruções no qual cada instrução será executada em sequência, como já fizemos em nossas demonstrações prévias. **Por exemplo**:

-  Siga em frente;
-  Vire à esquerda;
-  Vire à direita.

ou


<center><b>Figura 3:</b> Representação de estrutura sequencial.</center>

ou, em pseudocódigo: 

1. `a = leia_um_numero_do_teclado()`<br>
2. `b = leia_um_numero_do_teclado()`<br>
3. `c = leia_um_numero_do_teclado()`<br>
4. `media = ( a + b + c ) / 3`<br>
5. `exiba_media()`<br>

Como é sequencial, a sua ordem é importante. No exemplo anterior, não tem como calcularmos a média sem primeiro receber os valores de a, b e c.

### 2.2 Estruturas condicionais

Elas possibilitam a escolha de um grupo de ações e estruturas a serem executadas quando determinadas condições são ou deixam de ser satisfeitas. **Por exemplo**: 

-  Se a via estiver livre, acelere, caso contrário permaneça parado.

-  As estruturas condicionais acontecem quando a condição é verdadeira
-  Uma condição só pode ser **verdadeira** ou **falsa**

ou


<center><b>Figura 4:</b> Representação de estrutura condicional.</center>

ou, em pseudocódigo:

`se ( condição ) então:
    <instruções>`
    <br>
`senão:
    <instruções>`

Observe como se a condição for verdadeira, o caminho que você irá seguir será o da esquerda. Se a condição for falsa, ou seja, ela não ocorre, o caminho que você irá seguir é outro.

### 2.3 Estruturas de repetição

Elas possibilitam que um grupo de ações e estruturas sejam executadas enquanto (ou até que) determinadas condições deixam de ser satisfeitas. **Por exemplo**:


<center><b>Figura 5:</b> Representação de estrutura de repetição.</center>

#### Exemplo 4:

Esse algoritmo nos permite apresentar (imprimir na tela) todos os números de 1 até 20:

`numero = 1`
<br>
`enquanto ( numero <= 20 ) faça:
    exiba numero
    numero = numero + 1`
    
Neste exemplo começamos a contar com o valor de numero em 1. Como queremos contar até 20, temos que mostrar os números de 1 até 20.<br><br>
Partindo do número 1, fazemos uma pergunta: o valor de numero é menor que ou igual a 20? **Sim**. Então, exibimos o valor de numero (que é 1), e guardamos o próximo valor, que será o valor da variável numero mais 1, ou seja, 2. Fizemos nossas instruções e partimos novamente para nossa condição e repetimos nossa pergunta: o valor da variável numero é menor que ou igual a 20? A resposta continua sim! Então entramos novamente no nosso conjunto de instruções, exibimos o valor da variável e acrescentamos 1.<br><br>
Continuamos fazendo isto até que o valor de numero seja igual a 20. Quando ele entrar novamente, ele irá exibir o valor da variável numero (que será 20) e adicionar 1, tornando numero igual a 21. Quando for fazer a pergunta: o valor da variável numero é menor que ou igual a 20? A resposta será **não** e vamos parar de executar essas instruções. Contamos de 1 até 20.

Exemplo de um algoritmo que soma todos os números entre 200 e 400

 `soma = 0`

`enquanto ( num <= 400 ) faça:
    soma = soma + num

## 3. Linguagens de programação

Para todos os tipos de comunicação são necessárias linguagens. Os computadores apenas são capazes de entender código de máquina, que é de difícil interpretação pelos seres humanos. Dentro das linguagens de alto nível, podemos ter linguagens que são interpretadas e outras que são compiladas. Um código que é compilado necessita ser todo traduzido para código de máquina antes de sua execução, enquanto as interpretadas são convertidas linha por linha.

Vale ressaltar que em alguns exercícios a nota do site não vai condizer 100% com sua nota real, pois seu código ainda pode ser avaliado pelos monitores.

Para ter a sua nota e certificação **é imprescindível se cadastrar e fazer os workings, eles são parte considerável da sua nota.**

Para acessar o site basta ir em <a href="https://run.codes/">https://run.codes</a> e seguir o seguinte passo a passo:


1. Cliquem em "CADASTRE-SE AGORA!"
2. Preencha seus dados 
3. Clique em "CADASTRAR"
4. Vá ao seu email e confirme o cadastro
5. No campo "Universidade" coloque "PET-UFES-PET-Universidade Federal do Espírito Sando"
6. No campo "Num. Matrícula" coloque seu número de inscrição do introcomp.
7. No "Código de Matrícula" use o código que está sendo/foi dado em aula. Esse código identifica a turma e turno de vocês.
8. Clicando em "Página de Disciplina" você será redirecionado à sua turma e poderá a enviar seus códigos para avaliação.

Os cadeados verdes indicam que as questões já estão abertas e que você pode submeter o seu código fonte. Para enviar o código clique em "Ver Detalhes" e, em seguida, clique em "Selecionar Arquivo" e confirme.

**Ok, mas como deve estar o meu código?**

Os códigos devem estar escritos em Python e contidos em um arquivo com a terminação `.py`. Ele pode ser escrito em alguma IDE (como PyCharm) ou em algum editor de textos específicos (como NotePad++). Normalmente o desenvolvimento de códigos se dá de forma mais fácil em um ambiente Linux, porém não é necessário. 

**Quantas vezes posso enviar a mesma questão?**

Você pode enviar respostas para cada questão quantas vezes quiser enquanto ela estiver aberta. Nós avisaremos quando as questões serão abertas ou fechadas. Apenas o último envio da questão é considerado para fins avaliativos.

**Como funciona a correção?**

A correção acontece de forma automática. A plataforma executa o seu código e usa algumas entradas que nós fornecemos e compara com a saída esperada. Esse processo normalmente demora apenas alguns segundos. Fique a tento a isto. Se a saída esperada apresenta apenas letras minúsculas e você colocar maiúsculas, o resultado pode ser de errado. Apesar das respostas parecerem iguais, elas não seram para o run.codes. Fique atento com como a sáida esperada foi escrita.
