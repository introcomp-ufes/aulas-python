### 2.1 Janelas

A prxima coisa a se fazer  criar uma janela para o jogo. Para isso, vamos criar uma classe "Window, que tem como atributos: a altura (height) e a largura (width), alm da cor de fundo da tela (color).

```python
class Window:
    __width = 0 #largura
    __height = 0 #altura
    __color = (0, 0, 0) #cor

    def __init__(self, largura, altura): #construtor
        self.__width = largura
        self.__height = altura

    def returnWinSize(self): #retorna tamanho da tela
        return (self.__width, self.__height)

    def returnColor(self): #retorna a cor da janela
        return self.__color
```

### 2.4 Game Loop

O conceito de loop  bem recorrente em programao. No desenvolvimento de jogos, em particular, o conceito de **game loop**  a base para definir a sequencia de comportamentos derivados da interao entre jogador e jogo.

De maneira sucinta, dividimos o **game loop** em trs partes que se repetem em tempo de execuo (ilustrado na Figura 1):
- Input processing (processamento de entrada)
- Game update  (atualizao do jogo)
- Render (renderizao)


<center><b>Figura 1:</b> fluxograma de um game-loop.</center><br>

A parte de processamento de entrada  tambm conhecida como processamento de eventos.  a parte responsvel por detectar eventos advindos de dispositivo de input externos ao jogo como mouse, teclado, joystick etc.

J a parte de atualizao  responsvel por mudar o estado dos componentes do jogo de acordo com o processamento de eventos detectados na entrada. Atravs de linhas de cdigo,  decidido o que cada tipo de entrada encadeia dentro do nosso jogo, como por exemplo, apertar a barra de espao e executar um pulo.

Por fim, temos a renderizao. Ao final do game loop, com todas alteraes dentro do jogo feitas, ocorre a renderizao do nosso jogo, ou seja, sua construo visual.

 vlido pontuar que estes acontecimentos ocorrem em tempo de execuo, portanto,  interessante nos atentar ao quo rpido essas coisas acontecem.

Dentro de um gameloop, tambm precisamos dizer um tempo mnimo para o programa rodar cada iterao: se voc roda o mesmo jogo em um computador antigo e um computador de ponta, o que acontece  que a velocidade do jogo ficar muito mais baixa no computador antigo, e por mais que ambos consigam processar os dados em um tempo razoavel, isso gera inconsistncias. 

Podemos resolver esse problema utilizando o objeto `Clock`, presente no mdulo `time`.

<hr style="border-top: 1px dashed #234B6B; background-color: #E9B74F"/>

#### 2.4.1 Mdulo time

A biblioteca possui um mdulo capaz de monitorar o tempo, conhecida como `time` (tempo).

A partir deste mdulo, podemos pegar o tempo exato em um determinado momento do funcionamento do nosso jogo, pausar os processos fazendo o jogo esperar uma determinada quantidade de tempo e at mesmo criar sequncias de eventos. Voc pode encontrar a documentao deste mdulo no seguinte <a href="https://www.pygame.org/docs/ref/time.html#pygame.time.delay">link</a>.

<hr style="border-top: 1px dashed #234B6B; background-color: #E9B74F"/>

#### 2.4.2 Objeto Clock

O objeto Clock est presente no mdulo time e funciona como uma maneira mais elaborada de controlar o tempo, por meio de mtodos especficos para controlar a taxa de quadros por segundo (FPS - frames per second) que nosso jogo ir apresentar, ou seja, a frequncia com que o jogo ir executar suas funes.

Primeiro instanciamos o objeto antes de entrarmos no game-loop, e ento chamamos sua funo `tick` em seu interior, passando como parmetro a quantidade de FPS que desejamos limit-lo.

```python
#criamos uma instncia do relgio
clock = pygame.time.Clock()
count = 0

#inicio do game-loop
while True:
    #chamamos o tick do relgio para 30 fps e armazenamos o delta de tempo (diferena de tempo entre um ciclo e outro)
    delta = clock.tick(30)
    count += 1
    print(delta)
    
    if count == 5:
        break
```

Devemos chamar a funo `tick` a cada ciclo e, caso o ciclo anterior tenha sido mais rpido, tal funo suspende a execuo do nosso jogo por um breve tempo, mantendo assim, a freqencia que desejamos. Por fim, a funo retorna a diferena entre o tempo de execuo dos ciclos (delta).

Dessa forma, passa a ser possvel saber quanta coisa conseguimos processar no intervalo de um frame para outro dado os requisitos mnimos para seu jogo, evitando assim, as inconsistncias e problemas mencionados antes.

#### 2.5.1 Mdulo event

O mdulo `event`  o mdulo responsvel por gerenciar e lidar com eventos e filas de eventos.

Todas as teclas de dispositivos de entrada que forem apertadas entraro para uma fila de eventos a serem processados pelo computador. A funo `pygame.event.get()` nos d uma lista com esses eventos. Dessa forma, podemos iterar sobre essa fila, decidindo o que fazer ou como lidar com cada evento listado. O cdigo a seguir, por exemplo, l os eventos da lista de eventos em `pygame.event.get()` e trata trs tipos de eventos:

```python
run = True

while run:
    for event in pygame.event.get():
            if event.type == pygame.QUIT: #boto "x" utilizado para fechar janela
                run = False
            if event.type == pygame.KEYDOWN: #verifica se uma tecla foi pressionada
                if event.key == pygame.K_ESCAPE: #caso a tecla "esc" do teclado tenha sido a tecla pressionada, encerramos o loop/jogo
                    run = False
            
```

Aqui veremos a implementao das funes que iro definir o funcionamento do jogo. Algumas funes esto relacionadas  parte grfica do jogo da velha e, outras, relacionadas  prpria lgica do jogo. As subsees a seguir explicam as principais funes.

#### 3.2.2 Funo `draw_symbol()`

Nossa prxima funo ser responsvel por desenhar os smbolos <b>X</b> ou <b>O</b> no nosso tabuleiro. Na verdade no iremos exatamente desenhar estes smbolos, apesar de j termos conhecimento para isso. Nessa funo ns usaremos imagens prontas e as trataremos para que se encaixem no nosso tabuleiro.

```python
#funo para desenhar os simbolos no tabuleiro
def draw_symbol(position):
    global TURN
    x, y = position
    if TURN == 'PLAYER1':
        img = pygame.image.load('x.png').convert_alpha() #carregamos a imagem do nosso computador e damos transparncia a seus pixels.
        imgR = pygame.transform.scale(img, (100, 100)) #mudamos a escala da imagem para caber nas lacunas do tabuleiro
        tela.blit(imgR, (x - 50, y - 50)) #"printamos" a imagem na posio que queremos
        
    elif TURN == 'PLAYER2':
        img = pygame.image.load('o.png').convert_alpha()
        imgR = pygame.transform.scale(img, (100, 100))
        tela.blit(imgR, (x - 50, y - 50))
```

Aqui, utilizamos certas funes que no foram mostradas nesta aula por se trataram de um contedo um pouco mais avanado. Essas funes so utilizadas para o tratamento de imagens e voc pode conferir suas documentaes e conhecer melhor seus mdulos nos seguintes links:
- <a href="https://www.pygame.org/docs/ref/image.html">Mdulo image</a>.
- <a href="https://www.pygame.org/docs/ref/transform.html">Mdulo transform</a>.

A primeira funo que nos deparamos  a funo `load()` do mdulo `image`. Ela  responsvel por carregar uma imagem de um diretrio de nosso computador. Para isso passamos o mesmo como parmetro da funo. J o mtodo `convert_alpha()`  utilizado para dar transparncia ao fundo da imagem, como em imagens <b>.png</b>.

```python
#funcao para verificar a posio do click
def test_position():
    for rect in big_board:
        if e.type == pygame.MOUSEBUTTONDOWN and rect.collidepoint(mouse_pos):
            if rect == rect1:
                confirm(0, [100, 100])
            if rect == rect2:
                confirm(1, [300, 100])
            if rect == rect3:
                confirm(2, [500, 100])
            if rect == rect4:
                confirm(3, [100, 300])
            if rect == rect5:
                confirm(4, [300, 300])
            if rect == rect6:
                confirm(5, [500, 300])
            if rect == rect7:
                confirm(6, [100, 500])
            if rect == rect8:
                confirm(7, [300, 500])
            if rect == rect9:
                confirm(8, [500, 500])
```

A partir de um loop, passamos por cada retngulo da nossa varivel global `big_board`, verificamos se o boto do mouse foi pressionado e, utilizando a funo `collidepoint()`, se a posio do mouse na hora do click foi dentro do retngulo em questo. Essa ltima funo retorna um booleano True caso a posio passada para ela esteja dentro do retngulo e False caso contrrio. #### 3.2.4 Funo `confirm()`
Essa funo ser responsvel por verificar se a lacuna escolhida j foi preenchida ou no, verificando assim, se a jogada  vlida ou no.

```python
def confirm(index, position):
    global TURN, SYMBOL, COUNT
    
    if BOARD[index] == '-':
        
        BOARD[index] = SYMBOL
        draw_symbol(position)
        
        if TURN == 'PLAYER1':
            TURN = 'PLAYER2'
        else:
            TURN = 'PLAYER1'

        COUNT += 1
```

O funcionamento de tal funo  de fato bem simples. Utilizando nossa variveis globais, verificamos a possibilidade de preenchimento de uma lacuna. Se sim, desenhamos o smbolo e trocamos a vez do jogador, caso a jogada seja validada. Por fim, aumentamos em 1 nosso contador, que servir para definir um empate, caso o nmero mximo de jogadas seja atingido (9).

Passamos para a funo a posio na lista referente ao retngulo onde o click do mouse foi feito, dessa forma podemos alterar a mesma posio na nossa varivel global `BOARD`, deixando a mesma posio bloqueada para uma prxima jogada. Passamos tambm uma posio referente ao retngulo, para sabermos onde o smbolo dever ser desenhado.

#### 3.2.6 Funo ` text_victory()`

Esta funo ser responsvel por mostrar um texto em nossa tela anunciando o vencedor da partida ou ento, o empate do jogo.

```python
#funo para escrever o resultado da partida na tela
def text_victory(symbol):
    arial = pygame.font.SysFont('arial', 50)
    msg = 'JOGADOR {} VENCEU'.format(symbol)

    if symbol == 'EMPATE':
        msg_victory = arial.render('DEU VELHA!', True, (0, 0, 255), 0)
        superfRect = msg_victory.get_rect()
        superfRect.center = ((300), (300))
        tela.blit(msg_victory, superfRect)
    else:
        msg_victory = arial.render(msg, True, (0, 255, 0), 0)
        superfRect = msg_victory.get_rect()
        superfRect.center = ((300), (300))
        tela.blit(msg_victory, superfRect)
```

#### 3.2.7 Funo `reset()`

Por fim, temos a funo responsvel por resetar o jogo ao final de cada partida.

```python
#funo para resetar o jogo
def reset():
    global BOARD, STATE, TURN, SYMBOL, COUNT
    BOARD = [
        '-', '-', '-',
        '-', '-', '-',
        '-', '-', '-'
    ]
    STATE = 'PLAYING' 
    TURN = 'PLAYER1' 
    SYMBOL = 'X' 
    COUNT = 0 
    tela.fill((255,255,255))
```

A funo apenas seta os valores das variveis globais de volta para seus valores inicias e por fim preenche a tela com a cor branca.

### 3.3 Game loop e tratamento de eventos

Agora com todas as funes j definidas, inicializamos o mdulo Pygame, nossa janela de exibio e, ento, partimos para nosso `game loop`, onde trataremos os eventos que faro nosso jogo caminhar.

```python
#incializamos o mdulo pygame
pygame.init()

#inicializamos aqui nossa tela de exibio

janela = Window(600,600) #criamos nosso objeto janela (600x600)
tela = pygame.display.set_mode(janela.returnWinSize()) #inicializamos nossa janela
tela.fill(janela.returnColor()) #definimos a cor da janela 
pygame.display.set_caption('TicTacToe') #definimos o titulo da janela               
```

```python
#aqui iniciamos nosso game loop

while True:

    mouse_pos = pygame.mouse.get_pos() #pega a posio do mouse
    
    if STATE == 'PLAYING':
        
        draw_board() #desenha o tabuleiro

        for e in pygame.event.get(): #pega os eventos numa fila de eventos
            
            if e.type == pygame.QUIT: #caso o "x" da janela seja pressionado
                pygame.quit()

            if e.type == pygame.MOUSEBUTTONDOWN: #caso o boto do mouse seja pressionado
                if TURN == 'PLAYER1':
                    SYMBOL = 'X'
                    test_position()
                else:
                    SYMBOL = 'O'
                    test_position()    
            
            if test_victory('X'):
                text_victory('X')
                STATE = 'RESET' #ao final de uma partida, o jogo  resetado

            elif test_victory('O'):
                text_victory('O')
                STATE = 'RESET'

            elif COUNT >= 9:
                text_victory('EMPATE')    
                STATE = 'RESET'

    else:
        for e in pygame.event.get():
            if e.type == pygame.QUIT:
                pygame.quit()
               
            if e.type == pygame.MOUSEBUTTONDOWN:
                reset()
                draw_board()

    pygame.display.flip()    
```

Entendendo o funcionamento das funes, se torna simples entender a lgica por trs desse game loop.

Temos como novidade a funo `get_pos()`, que  responsvel por nos dar a posio do mouse durante todo o jogo. Guardaremos essa posio na varivel `mouse_pos`, que j foi vista anteriormente na funo <b>test_position()</b>. Essa funo faz parte do mdulo `mouse`, responsvel por trabalhar com a entrada gerada pelo nosso mouse.

Por fim, temos a funo `flip()` do nosso conhecido mdulo <b>display</b>. Tal funo  responsvel por atualizar toda nossa tela de exibio, diferentemente  da funo `update()` que tambm j conhecemos, uma vez que esta ltima atualiza apenas uma determinada parte da nossa tela.

Dessa forma, temos nosso primeiro jogo em pygame funcionando! Divirta-se!
