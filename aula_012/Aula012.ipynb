## 1.1 Contexto Histórico

A biblioteca teve seu desenvolvimento iniciado nos anos 2000 por Pete Shinners que, ao se familiarizar melhor com a linguagem Python e a biblioteca SDL (Simple DirectMedia Layer), decidiu unir essas duas coisas com o intuito de “fazer as coisas simples de maneira fácil e as coisas difíceis de maneira direta.”

A biblioteca SDL é escrita em linguagem C e é utilizada para facilitar o gerenciamento de recursos de multimídia do computador como por exemplo, sons, gráficos e dispositivos de entrada/saída. Estes recursos podem ser utilizados tanto por aplicações open sources, como por aplicações comerciais e empresariais.

A descontinuidade da implementação de tal biblioteca para python, criada por Mark Baker e conhecida como pySDL, foi umas das principais inspirações para Pete iniciar um projeto mais forte e robusto, sob a SDL, que mais tarde viria a ser chamado de Pygame.

As vantagens da PyGame se estendem das vantagens oferecidas pela própria linguagem e podem ser resumidas na palavra “simplicidade”. Além disso, a biblioteca possui uma boa documentação e conta com um grande apoio da comunidade, o que mantém o conteúdo sempre atualizado e com o mínimo de erros e “bugs”.

Já as desvantagens da biblioteca são, de certa forma, um reflexo dessa simplicidade, uma vez que o Python possui uma performance relativamente baixa, resolvendo em sua maioria, apenas gráficos 2D, não havendo um suporte a aceleração de hardware. 

Apesar disso, a biblioteca se mostra como uma excelente ferramenta de aprendizado na área de desenvolvimento de jogos e aplicações multimídia, cujo o conhecimento adquirido pode ser levado adiante, servindo como uma boa base para se trabalhar e aprender mais dentro de tal área.

### 2.1 Janelas

A próxima coisa a se fazer é criar uma janela para o jogo. Para isso, vamos criar uma classe "Window”, que tem como atributos: a altura (height) e a largura (width), além da cor de fundo da tela (color).

```python
class Window:
    __width = 0 #largura
    __height = 0 #altura
    __color = (0, 0, 0) #cor

    def __init__(self, largura, altura): #construtor
        self.__width = largura
        self.__height = altura

    def returnWinSize(self): #retorna tamanho da tela
        return (self.__width, self.__height)

    def returnColor(self): #retorna a cor da janela
        return self.__color
```

### 2.4 Game Loop

O conceito de loop é bem recorrente em programação. No desenvolvimento de jogos, em particular, o conceito de **game loop** é a base para definir a sequencia de comportamentos derivados da interação entre jogador e jogo.

De maneira sucinta, dividimos o **game loop** em três partes que se repetem em tempo de execução (ilustrado na Figura 1):
- Input processing (processamento de entrada)
- Game update  (atualização do jogo)
- Render (renderização)


<center><b>Figura 1:</b> fluxograma de um game-loop.</center><br>

A parte de processamento de entrada é também conhecida como processamento de eventos. É a parte responsável por detectar eventos advindos de dispositivo de input externos ao jogo como mouse, teclado, joystick etc.

Já a parte de atualização é responsável por mudar o estado dos componentes do jogo de acordo com o processamento de eventos detectados na entrada. Através de linhas de código, é decidido o que cada tipo de entrada encadeia dentro do nosso jogo, como por exemplo, apertar a barra de espaço e executar um pulo.

Por fim, temos a renderização. Ao final do game loop, com todas alterações dentro do jogo feitas, ocorre a renderização do nosso jogo, ou seja, sua construção visual.

É válido pontuar que estes acontecimentos ocorrem em tempo de execução, portanto, é interessante nos atentar ao quão rápido essas coisas acontecem.

Dentro de um gameloop, também precisamos dizer um tempo mínimo para o programa rodar cada iteração: se você roda o mesmo jogo em um computador antigo e um computador de ponta, o que acontece é que a velocidade do jogo ficará muito mais baixa no computador antigo, e por mais que ambos consigam processar os dados em um tempo razoavel, isso gera inconsistências. 

Podemos resolver esse problema utilizando o objeto `Clock`, presente no módulo `time`.

<hr style="border-top: 1px dashed #234B6B; background-color: #E9B74F"/>

#### 2.4.1 Módulo time

A biblioteca possui um módulo capaz de monitorar o tempo, conhecida como `time` (tempo).

A partir deste módulo, podemos pegar o tempo exato em um determinado momento do funcionamento do nosso jogo, pausar os processos fazendo o jogo esperar uma determinada quantidade de tempo e até mesmo criar sequências de eventos. Você pode encontrar a documentação deste módulo no seguinte <a href="https://www.pygame.org/docs/ref/time.html#pygame.time.delay">link</a>.

<hr style="border-top: 1px dashed #234B6B; background-color: #E9B74F"/>

#### 2.4.2 Objeto Clock

O objeto Clock está presente no módulo time e funciona como uma maneira mais elaborada de controlar o tempo, por meio de métodos específicos para controlar a taxa de quadros por segundo (FPS - frames per second) que nosso jogo irá apresentar, ou seja, a frequência com que o jogo irá executar suas funções.

Primeiro instanciamos o objeto antes de entrarmos no game-loop, e então chamamos sua função `tick` em seu interior, passando como parâmetro a quantidade de FPS que desejamos limitá-lo.

```python
#criamos uma instância do relógio
clock = pygame.time.Clock()
count = 0

#inicio do game-loop
while True:
    #chamamos o tick do relógio para 30 fps e armazenamos o delta de tempo (diferença de tempo entre um ciclo e outro)
    delta = clock.tick(30)
    count += 1
    print(delta)
    
    if count == 5:
        break
```

Devemos chamar a função `tick` a cada ciclo e, caso o ciclo anterior tenha sido mais rápido, tal função suspende a execução do nosso jogo por um breve tempo, mantendo assim, a freqûencia que desejamos. Por fim, a função retorna a diferença entre o tempo de execução dos ciclos (delta).

Dessa forma, passa a ser possível saber quanta coisa conseguimos processar no intervalo de um frame para outro dado os requisitos mínimos para seu jogo, evitando assim, as inconsistências e problemas mencionados antes.

#### 2.5.1 Módulo event

O módulo `event` é o módulo responsável por gerenciar e lidar com eventos e filas de eventos.

Todas as teclas de dispositivos de entrada que forem apertadas entrarão para uma fila de eventos a serem processados pelo computador. A função `pygame.event.get()` nos dá uma lista com esses eventos. Dessa forma, podemos iterar sobre essa fila, decidindo o que fazer ou como lidar com cada evento listado. O código a seguir, por exemplo, lê os eventos da lista de eventos em `pygame.event.get()` e trata três tipos de eventos:

```python
run = True

while run:
    for event in pygame.event.get():
            if event.type == pygame.QUIT: #botão "x" utilizado para fechar janela
                run = False
            if event.type == pygame.KEYDOWN: #verifica se uma tecla foi pressionada
                if event.key == pygame.K_ESCAPE: #caso a tecla "esc" do teclado tenha sido a tecla pressionada, encerramos o loop/jogo
                    run = False
            
```

K_UP = cima (↑)<br>

```python
superf = objeto.render(text, antialias, cor)
```

Tal função cria uma superfície e desenha nosso texto previamente configurado nela. Passamos para essa função, nosso texto, um booleano responsável por aplicar ou não o antialias (anti serrilhado) , e a cor que desejamos aplicar ao nosso texto (utilizando código RGB).

Agora precisamos desenhar essa superficie sobre nossa janela. É importante mencionar que ao fazer `objeto.render(...)` além da superficie, criamos também um retângulo que representa a área ocupada pelo texto. Para imprimir o texto, usaremos a função `.blit()`, colocando antes do ponto a janela criada.

Para conseguir o retângulo que envolve o texto, basta escrever `superf.get_rect()`. Além disso, podemos escolher onde o texto será colocado na tela movendo a posição deste retângulo. Para acessar o centro do retângulo, escrevemos:

```python
superfRect = superf.get_rect()
superfRect.center = ((250), (250))

win.blit(superf, superfRect) #lembrando sempre de chamar pygame.display.update(), caso contrario, a tela não será atualizada
```

### 3.1 Configurações Iniciais

Para começar, seguimos o que já aprendemos: importamos a biblioteca para nosso arquivo onde escreveremos nosso jogo e então criamos nossa classe janela, que trará as pré-configurações da nossa tela de exibição.

```python
#import da biblioteca
import pygame

#primeiro criamos um classe janela onde iremos pré-configurar nossa tela de exibição
class Window:
    __width = 0 #a definir
    __height = 0 #a definir
    __color = (255,255,255) #branco

    def __init__(self, largura, altura): #construtor
        self.__width = largura
        self.__height = altura

    def returnWinSize(self): #retorna tamanho da tela
        return (self.__width, self.__height)

    def returnColor(self): #retorna a cor da janela
        return self.__color
```

### 3.2 Funções

Aqui veremos a implementação das funções que irão definir o funcionamento do jogo. Algumas funções estão relacionadas à parte gráfica do jogo da velha e, outras, relacionadas à própria lógica do jogo. As subseções a seguir explicam as principais funções.   

Importante alertar que iremos utilizar variáveis globais que irão facilitar a programação do jogo. Portanto vamos definí-las antes de estudarmos as funções e entendermos seu real uso.

```python
#agora definimos algumas variáveis globais que serão responsáveis pelo funcionamento do jogo       

BOARD = [
    '-', '-', '-',
    '-', '-', '-',
    '-', '-', '-'
] #tabuleiro do jogo para verificação
STATE = 'PLAYING' #estado do jogo
TURN = 'PLAYER1' #vez de cada jogador
SYMBOL = 'X' #simbolo de cada jogador
COUNT = 0 #contador de jogadas

#definimos 9 retângulos que servirão de base para o tabuleiro
#cada retângulo representa uma lacuna do tabuleiro e uma parte da tela de exibição (600x600)
#[rect1|rect2|rect3]
#[rect4|rect5|rect6]
#[rect7|rect8|rect9]
rect1 = pygame.Rect((0, 0), (200, 200))
rect2 = pygame.Rect((200, 0), (200, 200))
rect3 = pygame.Rect((400, 0), (200, 200))
rect4 = pygame.Rect((0, 200), (200, 200))
rect5 = pygame.Rect((200, 200), (200, 200))
rect6 = pygame.Rect((400, 200), (200, 200))
rect7 = pygame.Rect((0, 400), (200, 200))
rect8 = pygame.Rect((200, 400), (200, 200))
rect9 = pygame.Rect((400, 400), (200, 200))

#definimos aqui, uma lista com os retângulos criados acima

big_board = [
    rect1,rect2,rect3,rect4,
    rect5,rect6,rect7,rect8,rect9,
] 
```

#### 3.2.1 Função `draw_board()`

Nossa primeira função será responsável por desenhar o nosso tabuleiro. Sabemos que este consiste em quatro linhas, duas na horizontal e duas na vertical. Lembrando que, no jogo da velha, estas linhas se interceptam formando um <b>#</b>.

```python
#aqui desenhamos nosso tabuleiro com linhas representando suas divisórias
def draw_board():
    pygame.draw.line(tela, (0, 0, 0), (200, 0), (200, 600), 10) 
    pygame.draw.line(tela, (0, 0, 0), (400, 0), (400, 600), 10)
    pygame.draw.line(tela, (0, 0, 0), (0, 200), (600, 200), 10)
    pygame.draw.line(tela, (0, 0, 0), (0, 400), (600, 400), 10)
```

As posições em que as linhas serão desenhadas foram pensadas para um display de 600x600, ficando igualmente espaçadas. A cor que definimos nessa função para as linhas é a cor preta e terão uma espessura de 10px. Já a superfície em que serão desenhadas é a nossa própria janela ou tela (que ainda não inicializamos).

Caso tenha alguma dúvida sobre os parâmetros dessa função, você pode voltar na seção <b>2.3 Linhas, retângulos e outros polígonos</b>.

 <hr style="border-top: 1px dashed #234B6B; background-color: #E9B74F"/>

#### 3.2.2 Função `draw_symbol()`

Nossa próxima função será responsável por desenhar os símbolos <b>X</b> ou <b>O</b> no nosso tabuleiro. Na verdade não iremos exatamente desenhar estes símbolos, apesar de já termos conhecimento para isso. Nessa função nós usaremos imagens prontas e as trataremos para que se encaixem no nosso tabuleiro.

```python
#função para desenhar os simbolos no tabuleiro
def draw_symbol(position):
    global TURN
    x, y = position
    if TURN == 'PLAYER1':
        img = pygame.image.load('x.png').convert_alpha() #carregamos a imagem do nosso computador e damos transparência a seus pixels.
        imgR = pygame.transform.scale(img, (100, 100)) #mudamos a escala da imagem para caber nas lacunas do tabuleiro
        tela.blit(imgR, (x - 50, y - 50)) #"printamos" a imagem na posição que queremos
        
    elif TURN == 'PLAYER2':
        img = pygame.image.load('o.png').convert_alpha()
        imgR = pygame.transform.scale(img, (100, 100))
        tela.blit(imgR, (x - 50, y - 50))
```

Aqui, utilizamos certas funções que não foram mostradas nesta aula por se trataram de um conteúdo um pouco mais avançado. Essas funções são utilizadas para o tratamento de imagens e você pode conferir suas documentações e conhecer melhor seus módulos nos seguintes links:
- <a href="https://www.pygame.org/docs/ref/image.html">Módulo image</a>.
- <a href="https://www.pygame.org/docs/ref/transform.html">Módulo transform</a>.

A primeira função que nos deparamos é a função `load()` do módulo `image`. Ela é responsável por carregar uma imagem de um diretório de nosso computador. Para isso passamos o mesmo como parâmetro da função. Já o método `convert_alpha()` é utilizado para dar transparência ao fundo da imagem, como em imagens <b>.png</b>.

A próxima função é a `scale()` do módulo `transform`. Esta é responsável por redimensionar uma imagem, para tal, passamos nossa imagem já carregada e as novas dimensões desejadas.

Nossa função `draw_symbol()` irá receber a posição, em coordenadas cartesianas, em que o símbolo deverá ser desenhado. A partir da variável global `TURN`a função identifica de qual jogador é a vez e então "desenha" o símbolo respectivo.

 <hr style="border-top: 1px dashed #234B6B; background-color: #E9B74F"/>

#### 3.2.3 Função `test_position()`

A função da vez agora, será responsável por verificar em qual lacuna do nosso tabuleiro o click foi feito.

```python
#funcao para verificar a posição do click
def test_position():
    for rect in big_board:
        if e.type == pygame.MOUSEBUTTONDOWN and rect.collidepoint(mouse_pos):
            if rect == rect1:
                confirm(0, [100, 100])
            if rect == rect2:
                confirm(1, [300, 100])
            if rect == rect3:
                confirm(2, [500, 100])
            if rect == rect4:
                confirm(3, [100, 300])
            if rect == rect5:
                confirm(4, [300, 300])
            if rect == rect6:
                confirm(5, [500, 300])
            if rect == rect7:
                confirm(6, [100, 500])
            if rect == rect8:
                confirm(7, [300, 500])
            if rect == rect9:
                confirm(8, [500, 500])
```

A partir de um loop, passamos por cada retângulo da nossa variável global `big_board`, verificamos se o botão do mouse foi pressionado e, utilizando a função `collidepoint()`, se a posição do mouse na hora do click foi dentro do retângulo em questão. Essa última função retorna um booleano True caso a posição passada para ela esteja dentro do retângulo e False caso contrário. Mais para frente veremos de onde tiramos a variável `mouse_pos`, que guarda a posição do mouse.

Caso a condição seja cumprida, chamamos a `função confirm()` que explicaremos a seguir.

 <hr style="border-top: 1px dashed #234B6B; background-color: #E9B74F"/>

#### 3.2.4 Função `confirm()`
Essa função será responsável por verificar se a lacuna escolhida já foi preenchida ou não, verificando assim, se a jogada é válida ou não.

```python
def confirm(index, position):
    global TURN, SYMBOL, COUNT
    
    if BOARD[index] == '-':
        
        BOARD[index] = SYMBOL
        draw_symbol(position)
        
        if TURN == 'PLAYER1':
            TURN = 'PLAYER2'
        else:
            TURN = 'PLAYER1'

        COUNT += 1
```

O funcionamento de tal função é de fato bem simples. Utilizando nossa variáveis globais, verificamos a possibilidade de preenchimento de uma lacuna. Se sim, desenhamos o símbolo e trocamos a vez do jogador, caso a jogada seja validada. Por fim, aumentamos em 1 nosso contador, que servirá para definir um empate, caso o número máximo de jogadas seja atingido (9).

Passamos para a função a posição na lista referente ao retângulo onde o click do mouse foi feito, dessa forma podemos alterar a mesma posição na nossa variável global `BOARD`, deixando a mesma posição bloqueada para uma próxima jogada. Passamos também uma posição referente ao retângulo, para sabermos onde o símbolo deverá ser desenhado.

 <hr style="border-top: 1px dashed #234B6B; background-color: #E9B74F"/>

#### 3.2.5 Função `test_victory()`

Tal função será responsável por verificar se alguma das condições para vitória foi atingida, retornando um booleano.

```python
#função para verificar condições de vitória
def test_victory(symbol):
    return ((BOARD[0] == symbol and BOARD[1] == symbol and BOARD[2] == symbol) or
        (BOARD[3] == symbol and BOARD[4] == symbol and BOARD[5] == symbol) or
        (BOARD[6] == symbol and BOARD[7] == symbol and BOARD[8] == symbol) or
        (BOARD[0] == symbol and BOARD[3] == symbol and BOARD[6] == symbol) or
        (BOARD[1] == symbol and BOARD[4] == symbol and BOARD[7] == symbol) or
        (BOARD[2] == symbol and BOARD[5] == symbol and BOARD[8] == symbol) or
        (BOARD[0] == symbol and BOARD[4] == symbol and BOARD[8] == symbol) or
        (BOARD[2] == symbol and BOARD[4] == symbol and BOARD[6] == symbol))   
```

 <hr style="border-top: 1px dashed #234B6B; background-color: #E9B74F"/>

#### 3.2.6 Função ` text_victory()`

Esta função será responsável por mostrar um texto em nossa tela anunciando o vencedor da partida ou então, o empate do jogo.

```python
#função para escrever o resultado da partida na tela
def text_victory(symbol):
    arial = pygame.font.SysFont('arial', 50)
    msg = 'JOGADOR {} VENCEU'.format(symbol)

    if symbol == 'EMPATE':
        msg_victory = arial.render('DEU VELHA!', True, (0, 0, 255), 0)
        superfRect = msg_victory.get_rect()
        superfRect.center = ((300), (300))
        tela.blit(msg_victory, superfRect)
    else:
        msg_victory = arial.render(msg, True, (0, 255, 0), 0)
        superfRect = msg_victory.get_rect()
        superfRect.center = ((300), (300))
        tela.blit(msg_victory, superfRect)
```

Nesta função, utilizamos a função `SysFont()`, que é responsável por criar um objeto `Font` a partir das fontes presentes no sistema. Continuando, realizamos o que já aprendemos na seção <b>2.6 Textos e fontes</b>.

 <hr style="border-top: 1px dashed #234B6B; background-color: #E9B74F"/>

#### 3.2.7 Função `reset()`

Por fim, temos a função responsável por resetar o jogo ao final de cada partida.

```python
#função para resetar o jogo
def reset():
    global BOARD, STATE, TURN, SYMBOL, COUNT
    BOARD = [
        '-', '-', '-',
        '-', '-', '-',
        '-', '-', '-'
    ]
    STATE = 'PLAYING' 
    TURN = 'PLAYER1' 
    SYMBOL = 'X' 
    COUNT = 0 
    tela.fill((255,255,255))
```

A função apenas seta os valores das variáveis globais de volta para seus valores inicias e por fim preenche a tela com a cor branca.

### 3.3 Game loop e tratamento de eventos

Agora com todas as funções já definidas, inicializamos o módulo Pygame, nossa janela de exibição e, então, partimos para nosso `game loop`, onde trataremos os eventos que farão nosso jogo caminhar.

```python
#incializamos o módulo pygame
pygame.init()

#inicializamos aqui nossa tela de exibição

janela = Window(600,600) #criamos nosso objeto janela (600x600)
tela = pygame.display.set_mode(janela.returnWinSize()) #inicializamos nossa janela
tela.fill(janela.returnColor()) #definimos a cor da janela 
pygame.display.set_caption('TicTacToe') #definimos o titulo da janela               
```

```python
#aqui iniciamos nosso game loop

while True:

    mouse_pos = pygame.mouse.get_pos() #pega a posição do mouse
    
    if STATE == 'PLAYING':
        
        draw_board() #desenha o tabuleiro

        for e in pygame.event.get(): #pega os eventos numa fila de eventos
            
            if e.type == pygame.QUIT: #caso o "x" da janela seja pressionado
                pygame.quit()

            if e.type == pygame.MOUSEBUTTONDOWN: #caso o botão do mouse seja pressionado
                if TURN == 'PLAYER1':
                    SYMBOL = 'X'
                    test_position()
                else:
                    SYMBOL = 'O'
                    test_position()    
            
            if test_victory('X'):
                text_victory('X')
                STATE = 'RESET' #ao final de uma partida, o jogo é resetado

            elif test_victory('O'):
                text_victory('O')
                STATE = 'RESET'

            elif COUNT >= 9:
                text_victory('EMPATE')    
                STATE = 'RESET'

    else:
        for e in pygame.event.get():
            if e.type == pygame.QUIT:
                pygame.quit()
               
            if e.type == pygame.MOUSEBUTTONDOWN:
                reset()
                draw_board()

    pygame.display.flip()    
```

Entendendo o funcionamento das funções, se torna simples entender a lógica por trás desse game loop.

Temos como novidade a função `get_pos()`, que é responsável por nos dar a posição do mouse durante todo o jogo. Guardaremos essa posição na variável `mouse_pos`, que já foi vista anteriormente na função <b>test_position()</b>. Essa função faz parte do módulo `mouse`, responsável por trabalhar com a entrada gerada pelo nosso mouse.

Por fim, temos a função `flip()` do nosso conhecido módulo <b>display</b>. Tal função é responsável por atualizar toda nossa tela de exibição, diferentemente  da função `update()` que também já conhecemos, uma vez que esta última atualiza apenas uma determinada parte da nossa tela.

Dessa forma, temos nosso primeiro jogo em pygame funcionando! Divirta-se!
