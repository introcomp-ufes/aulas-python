## 1. Modularizao com arquivos

Refrescando um pouco a sua memria, h poucas aulas atrs te ensinamos uma forma de modularizar seu cdigo, por meio das funes. Entretanto, com o crescimento de um projeto, modularizar apenas por meio de funes pode no ser suficiente para ter um cdigo esteticamente agradvel e, principalmente, legvel.

A alternativa mais simples e popular para resolver esse problema  a separao do programa em <b>arquivos.</b> Com a separao inteligente do cdigo em arquivos, podemos agrupar funes e classes relacionadas  resoluo de problemas parecidos e, apenas referenciando um arquivo, us-las em um programa escrito em outro arquivo. Desta forma, podemos **testar** os mdulos de forma independente, ou seja, mais eficientemente. Como arquivos podem ser importados por outros arquivos e, consequentemente, funes e classes tambm so "importadas", estamos permitindo outra caracterstica desejvel em prograo, a saber, o **reuso** de cdigo.

Na teoria, parece um pouco complicado e muito abstrato, porm, na prtica, veremos que  bem simples!

### 1.1. Separao em arquivos

Abaixo temos um programa main.py que recebe um nmero inteiro e exibe algumas caractersticas relacionadas ao autor deste exemplo. Perceba que j temos algumas funes modularizando o cdigo:

````python
#main.py

#funes de matemtica
def paridade(numero):
    if numero%2 == 0:
        return "par"
    else:
        return "mpar"
        
def fatorial(numero):
    fatorial = 1
    
    while (numero > 1):
        fatorial = fatorial * numero
        numero = numero - 1
        
    return fatorial

#funes de datas
def eh_mes_do_aniversario(mes):
    if(mes == 11):
        return " o ms do aniversrio do criador do exemplo!"
    
    return "No  o ms que eu queria :("

def eh_dia_do_aniversario(dia):
    if(dia == 1):
        return " o dia do aniversrio do criador do exemplo!"
    
    return "No  o dia que eu queria :("

def eh_ano_de_nascimento(ano):
    if(ano == 1999):
        return " o ano do nascimento do criador do exemplo!"
    
    return "No  o ano que eu queria :("

#programa principal
numero = int(input())
print("O nmero  " + paridade(numero))
print("O fatorial  " + str(fatorial(numero)))
print(eh_dia_do_aniversario(numero))
print(eh_mes_do_aniversario(numero))
print(eh_ano_de_nascimento(numero))
````

Com o programa feito, podemos tentar agrupar as funes de forma inteligente. Preferimos separar as duas primeiras funes (relacionadas a matemticas) e as ltimas trs que lidam com data.

J com as categorias separadas, faremos um arquivo para cada uma delas e manteremos o programa principal com o mesmo nome.

#### 1.2.2. Importando mdulos

Chamamos aquilo que importamos de outro arquivo de mdulo, seja uma funo, classe ou o arquivo inteiro. No muito diferente da situao anterior, para importar funes e classes, devemos continuar utilizando a palavra reservada `import`, porm precedida de `from`.

````python
#main.py

from minhamatematica import paridade
from datasdocriador import eh_dia_do_aniversario, eh_mes_do_aniversario

numero = int(input())
print("O nmero  " + paridade(numero))
print(eh_dia_do_aniversario(numero))
print(eh_mes_do_aniversario(numero))
````

Nessa situao, retiramos algumas funcionalidades do programa principal para melhor aproveitarmos o conceito apresentado. Traduzindo `from arquivo import funo` literalmente para o portugus, teramos algo como `de arquivo importa funo`.

Com isso, apenas carregamos para o programa principal as funes que sero utilizadas, tendo um programa mais leve e legvel.

Essa sintaxe tambm pode ser utilizada para importar todas as funes de um arquivo, sem a necessidade de escrever todos os nomes. Para isso, basta escrever `from arquivo import *`. Apesar de funcionar,  pouco usado pela comunidade geral de Python, principalmente por perder o controle dos nomes de funes que j so utilizados.

Utiliza-se esta mesma sintaxe para importar classes de outros arquivos. Utilizando o arquivo usuario.py, que possui uma classe Usuario, observe o exemplo abaixo:

````python
from usuario import Usuario

usuario1 = Usuario("Chapolin", 20, "01/11")

print(usuario1.maiorDeIdade())
usuario1.imprimeAniversario()
````

O mais importante de ser absorvido deste caso  que basta importar a classe Usuario que automaticamente todos os seus mtodos tornam-se utilizveis no programa.

## 3. Manipulando matrizes com Numpy

Sabemos que Python possui suporte "nativo" a "matrizes", mas isso no funciona muito bem. O suporte, na verdade, no  nativo pois as matrizes em Python so definidas como listas de listas. Os problemas surgem quando queremos multiplicar uma matriz por um vetor, ou uma matriz por outra matriz. Nestes casos, dois problemas principais podem ocorrer: performance e praticidade. A performance  afetada pelo uso excessivo de loops; quanto  praticidade, no  nada amigvel multiplicar "matrizes" (ou seja, listas de listas) em Python.

Para contornar estes problemas, utilizaremos a bibloteca `numpy` que oferece inmeras ferramentas matemticas, inclusive, oferece apoio ao desenvolvimento de matrizes com sintaxe bastante amigvel. `numpy` ainda  otimizada para operaes com matrizes e, portanto, performance no ser mais um problema.

Para aqueles que no conhecem um vetor, iremos defin-lo como algo muito semelhante a uma lista. Pense num vetor $A = \begin{bmatrix} 1 & 2 & 3 & 4 \end{bmatrix}$. Muito semelhantes s listas que j conhecemos, certo? Se quisermos acessar o elemento de valor $2$ devemos acessar o segundo ndice do vetor, ou seja, `A[1]`, exatamente como fazemos com listas.

Uma matriz seria um vetor com vrios vetores. Imagine agora o seguinte vetor:

$M = \begin{bmatrix} M_{[0]} & M_{[1]} & M_{[2]} & M_{[3]} \end{bmatrix}$

Se `M[0]`, `M[1]`, `M[2]` e `M[3]` forem vetores, `M`  uma matriz. Imagine `M[0] = [1, 2, 3, 4]`, `M[1] = [5, 6, 7, 8]`, `M[2] = [9, 10, 11, 12]` e `M[3] = [13, 14, 15, 16]`. Temos, nesse caso:

$M = \begin{bmatrix} 1 & 5 & 9 & 13 \\
                     2 & 6 & 10 & 14 \\
                     3 & 7 & 11 & 15 \\
                     4 & 8 & 12 & 16
    \end{bmatrix}$
    
Se estivermos interessados no elemento que vale $10$ devemos utilizar a notao `M[2][1]` pois queremos o segundo elemento do terceiro vetor. De uma forma geral, temos a notao `M[i][j]` onde `i`  a linha e `j`, a coluna.
   
**Observao:** durante o ensino mdio  ensinado que o primeiro ndice de uma matriz  $1$, aqui, entretanto, utilizaremos o primeiro ndice como $0$ para simplificar o aprendizado da biblioteca.

O primeiro passo para usar tal biblioteca  import-la:

````python
import numpy as np
````

Nosso prximo passo  criar uma matriz. Existem diversas formas de fazer isso e veremos algumas delas a seguir.

### 3.1. Criando matrizes

Vamos dividir essa seo em casos; cada caso representa uma possvel situao em seu cdigo.

#### Matriz nula e similares:

Caso queira criar uma matriz nula de tamanho $m{\times}n$, basta chamar a funo `np.zeros(m, n)` [[ref]](https://numpy.org/doc/1.18/reference/generated/numpy.zeros.html)

````python
# Criando matriz nula de tamanho 3x4
matriz_nula = np.zeros((3, 4))
print(matriz_nula)
````

Repare que os zeros da nossa matriz so do tipo `float`; podemos definir o tipo dos nossos dados por meio do parmetro `dtype`. Desse modo:

````python
matriz_nula = np.zeros((3, 4), dtype=int)
print(matriz_nula)
````

Podemos tambm criar uma matriz cheia de `1`s chamando a funo `np.ones` [[ref]](https://docs.scipy.org/doc/numpy-1.14.1/reference/generated/numpy.ones.html)

````python
matriz_um = np.ones((3, 4), dtype=int)
print(matriz_um)
````

**NOTA:** Assim como na matemtica, em Python, os nomes de variveis que representam matrizes so normalmente uma letra mauscula. `matriz_nula` e `matriz_um` no so bons nomes para matrizes e, portanto, a partir de agora utilizaremos nomes como `A`, `B`, `M`, `I`, `X`, `Y` e `Z`.

A partir das matrizes que j sabemos criar, podemos criar muitas outras matrizes atravs de operaes bsicas de matrizes, como **multiplicao por escalar**, **soma por escalar** e **soma de matrizes**. Os trechos de cdigo a seguir mostram exemplos de manipulao de matrizes.

````python
# Obtendo uma matriz 3x3 cheia de 3:
A = np.ones((3, 3), dtype=int) * 3
print(A)
````

````python
# Obtendo uma matriz 3x3 cheia de 2:
B = np.zeros((3, 3), dtype=int) + 2
print(B)
````

````python
# Obtendo uma matriz 3x3 cheia de 5:
C = A + B
print(C)
````

Perceba como  simples manipular matrizes e como nos aproximamos muito da matemtica com `numpy`.

**Observao 1:** na matemtica, quando dizemos que somamos $B+2$, onde $B$  uma matrix $2{\times}2$, estamos na realidade fazendo 

<center>
$B_{2{\times}2}+\begin{bmatrix}2 & 2\\2 & 2\end{bmatrix} = \begin{bmatrix}B_{[0][0]} + 2 & B_{[0][1]} + 2\\B_{[1][0]} + 2 & B_{[1][1]} + 2\end{bmatrix}$
</center>
    
e  exatamente isso que ocorre em nosso cdigo. Pode no parecer til no momento mas, quando trabalhamos com Estatstica e _Machine Learning_, tal comportamento se torna um grande facilitador, uma vez que no precisamos recorrer a _loops_.

**Observao 2:** chamaremos qualquer _array_ de matriz ao longo dessa aula para melhorar a didtica; no necessariamente estaremos nos referindo a _arrays_ bidimensionais ao falarmos de matrizes.

Acabamos de ver uma biblioteca muito poderosa de Python, com altssimo nvel de aplicabilidade para programas com maior profundidade terica. Agora veremos a biblioteca Turtle que, apesar do seu teor ldico, vai nos ajudar a entender as propores do poder das bibliotecas em Python.

### 4.1. Conceitos bsicos

Referenciando os conceitos da aula anterior, a biblioteca funciona na sua maior parte com os mtodos de duas classes centrais, a `TurtleScreen`, que representa o display no qual as imagens sero exibidas, e a `RawTurtle`, uma seta exibida nos desenhos que desenha no display.

Apesar de toda a programao girar em torno dessas classes, os conceitos de orientao a objetos podem ser dispensveis. A biblioteca oferece uma "interface procedural" que contm funes de todos os mtodos dessas duas classes, e sempre que uma funo  chamada automaticamente so criados objetos das classes utilizadas. Na teoria esse conceito pode ser complicado, mas com alguns poucos exemplos torna-se intuitivo.

Para iniciar a demonstrao da biblioteca, podemos fazer o seguinte cdigo base:

````python
from turtle import *

forward(0)

done()
````

Iniciamos com a importao da biblioteca e logo escrevemos o primeiro comando, que cria automaticamente um objeto TurtleScreen para exibir um display e outro objeto RawTurtle que renderiza uma seta apontando para a direita. J o segundo comando nos permite fechar o display pelo "x" (se voc esquecer dele pode ficar meio perdido :P). Observe o resultado abaixo:


<center><b>Figura 02:</b> Resultado do cdigo acima.</center>

A seta ser o referencial do desenho e funciona analogamente como uma caneta encostando em um papel. De maneira bsica, podemos moviment-la para frente e para trs com forward() e backward(), respectivamente. Para usarmos essas funes, fornecemos um valor inteiro que representa o nmero de pixels a ser percorrido, seja para frente ou para trs. Veja o cdigo exemplo abaixo:

````python
from turtle import *

forward(250)

done()
````

Executando o cdigo, obtemos uma reta de 250 pixels com a seta apontando para a direita, como mostra a imagem abaixo:


<center><b>Figura 03:</b> Trao de 250 pixels para a direita.</center>

Para alterarmos a direo da nossa reta, podemos utilizar as funes `left()` e `right()`, que recebem um valor que representa um ngulo em graus em que a seta ir girar. Veja o cdigo abaixo e perceba o resultado:

````python
from turtle import *

forward(100)
left(90)
forward(100)

done()
````


<center><b>Figura 04:</b> Traos resultantes do cdigo acima.</center>

Dessa vez temos mais de uma funo no programa. A lgica das classes e objetos criados acaba sendo abstrada, de forma que os objetos necessrios so criados apenas na primeira funo e reutilizados em todas as outras seguintes.

Alm das movimentaes a partir dos ngulos e distncias, a biblioteca fornece uma funo muito mais simples para situaes em que se deseja mover a seta para uma posio especfica do display. Com `setposition()`, enxergamos o display como um plano cartesiano, onde a origem  o ponto central. A partir disso, podemos fornecer dois nmeros que levam a seta at um ponto do display:

````python
from turtle import *

#setposition(coordenada_x, coordenada_y)

setposition(100, 100)
setposition(-100, 100)
setposition(-100, -100)
setposition(100, -100)
setposition(100, 100)

done()
````


<center><b>Figura 05:</b> Figura formada utilizando setposition(). </center>

Lembra da analogia da caneta? Repare que em todas as movimentaes a seta sempre traou retas, como uma caneta presa em um papel. Para melhor controle do desenho, tambm temos as funes `penup()` e `pendown()`, que funcionam como "tirar a caneta do papel" e  "prender a caneta no papel", respectivamente. Adaptando o exemplo anterior, conseguimos formar um quadrado perfeito centralizado:

````python
from turtle import *

#aps esse comando, as movimentaes no formam mais desenhos
penup()
setposition(100, 100)
#com esse comando, as movimentaes voltam a desenhar
pendown()

setposition(-100, 100)
setposition(-100, -100)
setposition(100, -100)
setposition(100, 100)

done()
````


<center><b>Figura 06:</b> Quadrado com lado de 200 pxels. </center>

Outros comandos interessantes nos permitem preencher a imagem, traando uma reta do ponto em que o desenho se iniciou at o ltimo ponto desenhado, preenchendo a parte interior. Para isso, basta sinalizarmos o incio da parte a ser preenchida com `begin_fill()` e o fim com `end_fill()`: 

````python
from turtle import *

begin_fill()

forward(100)
left(90)
forward(100)

end_fill()

done()
````

### 4.2. Utilizando a biblioteca com loops

Agora que conhecemos os recursos bsicos, podemos explorar um pouco mais alguns desenhos bem divertidos e com poucas linhas de cdigo, utilizando principalmente loops.

#### Exemplo 1: Tringulo equiltero

Lembrando que a seta do desenho inicia-se apontando para a direita, basta escolhermos o tamanho do lado do tringulo e virar 120 para um dos lados e interromper o loop ao atingirmos novamente a posio central do display. Acompanhe o exemplo para ficar mais claro:

````python
from turtle import *

# esse comando determina a velocidade de desenho como 10 (velocidade mxima)
speed(10)

begin_fill()

while True:
    forward(100) ## tamanho do lado do triangulo 
    left(120) ## ngulo externo do tringulo
    
    coordenadas = pos()
    
    # verifica se a seta retornou ao ponto de origem
    if abs(coordenadas) < 1:
        break

end_fill()
    
done()
````


<center><b>Figura 08:</b> Tringulo equiltero.</center>

O ngulo de 120 foi escolhido justamente por ser o ngulo externo do tringulo equiltero. Assim, para desenhar qualquer polgono regular basta escolher o tamanho e calcular o ngulo externo.

#### Exerccio 5: Polgonos regulares

Faa um programa que recebe a quantidade de lados e desenhe o polgono regular correspondente de lado com tamanho de 100 pixels.

Frmulas que podem te ajudar: <br>
`ngulo externo = 180 - ngulo interno`<br>
`ngulo interno = [(lados - 2) * 180]/lados`

````python
from turtle import *

lados = int(input("Nmero de lados (deve ser maior que 2): "))

#ache a formula para encontrar o angulo a partir da quantidade de lados
angulo = 0

speed(10)

begin_fill()

while True:
    forward(100)
    left(angulo) 
    
    coordenadas = pos()

    if abs(coordenadas) < 1:
        break

end_fill()
    
done()
````

#### Exemplo 2: Cruzando linhas

Ao escolhermos um ngulo entre 180 e 270 para a esquerda (ou 90 e 180 para a direita), as linhas desenhadas se cruzaro. Dependendo do ngulo escolhido, pode-se formar imagens bem interessantes! Observe o exemplo abaixo:

````python
from turtle import *

speed(10)

while True:
    forward(150)
    left(220)
    
    coordenadas = pos()
    
    if abs(coordenadas) < 1:
        break
    
done()
````

## 5. O que pode dar errado?

### 5.1. Bibliotecas no instaladas

Imagine a seguinte situao: voc e seu amigo esto fazendo uma tarefa em Python, tal tarefa requer o uso da biblioteca hipottica `libcomp`. Ao finalizar a parte dele da tarefa seu amigo o envia o cdigo-fonte para que voc possa fazer a sua parte, entretanto voc no instalou a biblioteca `libcomp` e o interpretador exibiu uma mensagem de erro.

A situao apresentada  muito comum, principalmente quando trabalhamos com _scripts_ e programas com muitas dependncias. Uma maneira de resolver isso  criando um arquivo `requirementes.txt` contendo uma lista de dependncias do seu programa, desse modo qualquer um que quiser usar sua aplicao s precisar executar `pip install -r requirements.txt` e todas as dependncias ausentes sero automaticamente baixadas por meio do `pip`. No Windows  possvel performar a mesma ao com o `conda` atravs do comando `conda install --file requirements.txt`.

Perceba que um arquivo `requirements.txt` de um projeto que depende das bibliotecas `numpy`, `torch` e `cv2` :
```
numpy
torch
opencv-python
```
Existem formas mais avanadas de utilizar o `requirements.txt`, elas podem ser encontradas (em portugus) [aqui](https://jtemporal.com/requirements-txt/).
