## 3. Listas

Uma lista é uma conjunto de valores que são chamados de elementos ou itens da lista. As listas podem conter itens de qualquer tipo, inclusive conter outras listas. Cada item de uma lista é indexado, ou seja, tem um número atrelado à ele que indica a posição do valor na lista. Essa contagem começa com o primeiro item da lista tendo o índice `0`. Uma lista que tem uma ordem de seus itens é chamada de sequência.

Podemos utilizar o exemplo de um carrinho de compras para ilustrar a utilidade de uma lista, ao invés de usar um conjunto de variáveis: digamos que você esteja implementando um programa no qual deseja-se alguma forma de guardar itens em um carrinho de compras. Utilizando o nosso conhecimento atual poderíamos declarar algumas variáveis:

```python
item1 = "Tomates"
item2 = "Batatas"
item3 = "Ovos"
item4 = "Pães"
```

Vejamos como seria a impressão na tela dos itens do carrinho sem a utilização de listas:

```python
print(item1)
print(item2)
print(item3)
print(item4)
```

Mas, como podemos ver, teríamos que digitar e lembrar de cada uma dessas variáveis quando formos utilizá-las no programa; isso não é eficiente e dificulta na automatização de tarefas no programa. Ao invés de variáveis, vamos implementar uma lista com esses itens:

```python
carrinho = ["Tomates", "Batatas", "Ovos", "Pães"]
```

Muito mais simples e direto. Agora, para acessar os itens da lista, podemos referenciar os índices (ou seja, posições dos itens na lista). Então, caso precisemos imprimir o primeiro item da lista basta utilizar o índice `0`:

```python
print(carrinho[0])
```

### 3.6 Adicionar e remover itens

Podemos adicionar e remover itens de uma lista por meio de funções padrão de Python. Lembrando que uma das propriedades mais interessantes das listas é poder ter tamanho variável, ou seja, itens podem ser adicionados e removidos sob demanda! Veremos algumas destas funções nas seções a seguir.

#### 3.6.1 Adicionando itens com `append()`

#### 3.6.2 Removendo itens com `pop()` e `remove()`
As funções `pop()` e `remove()` permitem retirar itens de uma lista. A função `pop()` pode receber uma posição e o item é retirado daquela posição:

```python
print("Estou removendo o item: ", carrinho.pop(3) ) # Removendo item na posição 3
print("Carrinho atualizado:" , carrinho )
```

```python
print("Estou removendo o item: ", carrinho.pop() )
print("Carrinho atualizado:" , carrinho )
```

A função `remove()` busca pelo item dado e o remove da lista, caso ele esteja presente:

```python
if "Batatas" in carrinho:
    print("Estou removendo as batatas: ", carrinho.remove("Batatas") )
print("Carrinho atualizado:" , carrinho )
```

## 4. Loops
Um loop (ou laço, em português) nada mais é do que uma estrutura de código que permite repetir comandos por uma dada quantidade de vezes ou enquanto uma condição seja satisfeita. Eles são extremamente necessários pois permitem repetição de comandos. Vamos abordar dois tipos de loop nessa aula, o `while` e o `for`, que são os principais laços do Python. Um loop consiste basicamente de um conjunto de comandos aninhados e uma condição de parada, que diz até quando esses comandos devem ser repetidos.

### 4.1 Condição de parada e contadores

Os conceitos da aula sobre condicionais e operadores lógicos serão utilizados nos algoritmos que iremos estudar hoje pois os loops precisam saber até onde ou quando precisam ser executados.
    
Caso você peça para alguém fazer uma tarefa, provavelmente você também irá indicar um momento onde ela terá de parar de realizar tal tarefa. Essa condição também existe para os computadores e iremos chamá-la formalmente de **condição de parada**.

Um conceito que também utilizamos na programação é o de **contadores**. Contadores são variáveis numéricas e servem para contar algum processo repetitivo, saber a posição atual de algum programa, fazer testes de depuração ou apenas para que um laço aconteça determinada quantidade de vezes. ### 4.2 O Loop `while`

O loop `while` é uma estrutura de repetição na qual uma condição é verificada antes que os comandos dentro do loop sejam executados. Se a condição for verdadeira, um conjunto de comandos é executado e, caso contrário, os comandos não são executados. Assim que os comandos forem executados, a condição é avaliada novamente e todo o processo se repete.

```python
x = 0           # Inicializando x

print("Entrando no laço")

while x < 5 :   # Enquanto x for menor do que 5 execute:
    print(x)    # Imprima o valor de x
    x = x + 1   # Adicione 1 ao valor de x. Também pode ser feito como x += 1
    
print("Saímos do laço")
```

Um bom exercício é olhar o código e tentar prever o passo a passo das linhas e comandos que o computador vai executar. Também podemos combinar condições lógicas na chamada do `while`. Por exemplo, olhe o código abaixo e pense quantos números serão impressos na tela: 

```python
x = 0

while x < 100 and x != 20 :
    print("X está valendo " , x )
    x += 1

print("Saímos do laço")
```

### 4.3 O Loop `for`

O loop `for` tem a funcionalidade geral muito parecida com a de um loop `while`, porém, ele permite simplificar alguns tipos de laços específicos e é mais comumente utilizado. O `for` exige a pela criação de um contador e a definição da condição de parada na sua chamada. Veremos algumas formas de como chamar esse laço.

Podemos definir o `for` e passar um contador (ou iterador) e um domínio (ou alcance), como no exemplo a seguir:

```python
for x in range(0,5):   # Para "x" no intervalo de 0 à 5 (não inclusivo, intervalo aberto à direita)
    print(x)           # Imprima "x"
```

A função `range()` pode ter uma quantidade variável de parâmetros. Você pode, por exemplo, dar apenas o inteiro para que o `for` pare de percorrer:

```python
for contador in range(12): # Note que se o início não for dado, o contador começa pelo 0
    print(contador)
```

A função `range` pode receber o início e o fim, como no primeiro exemplo de laço `for`, ou também receber 3 parâmetros que definem: o início do laço, o fim e o quanto ele deve acrescentar no contador por cada iteração:

```python
for numero in range( 2 , 100, 5 ):
    print(numero)
```

Então já vimos algumas vantagens para essa implementação do `for` comparado ao `while`. Não precisamos criar o contador antes do laço e a função `range()` já toma conta da iteração sozinha. Mais uma coisa legal que podemos fazer como o `for` é percorrer listas! Olhem só:

```python
baralho = ["Ás" , "2" , "3" , "4" , "5" , "6" , "7" , "8" , "9" , "J" , "Q" , "K"  ]

for carta in baralho: # Para cada carta na lista baralho
    print( carta )    # Imprima a carta
    
```

De forma bem simples e direta pegamos um contador e fazer ele passar por todos os itens de uma lista individualmente a cada iteração. Fantástico, não? E esse arranjo serve para qualquer tipo de lista.

### 4.4 Desvios de fluxo

Existem também algumas formas de "desviar" o fluxo de um laço. Dois deles são o `break` e o `continue`. Por exemplo, considere a situação na qual seja necessário voltar ao começo do laço sem terminar de executar todo o bloco de código. Usaremos para isso o `continue`, como no exemplo a seguir:

```python
x = 0

while x < 12 :

    x += 1
    if ( x % 2 == 0 ):  #Fazendo um teste para saber a paridade de "x"
        continue        #Se "x" for par, volte para o início do laço
    
    print(x)
```

E no caso do `break`, podemos interromper completamente um laço, sair dele e continuar abaixo do bloco de código, como no exemplo a seguir:

```python
x = 0

while x < 99 :    #Laço de "x" enquanto "x" for menor que 99

    x += 1
    if x == 8 :   #Se "x" for igual a 8
        break     #Saia do laço
        
    print(x)

print("Saímos do laço")
```

## 5. Loops aninhados

Os loops aninhados nada mais são que sequências de loops definidos dentro de outros loops, ou seja, laços dentro de laços. Parece confuso, certo? Talvez um exemplo ajude a ilustrar melhor a ideia.<br>

Um exemplo que podemos citar está no nosso próprio calendário/relógio. Nosso ano possui 12 meses e se repete de janeiro a dezembro, caracterizando um laço. Dentro do nosso ano, temos nossos meses e dentro destes, temos os dias (mais um laço). Não obstante, ainda temos as horas, dentro dessas, os minutos, dentro desses, os segundos e assim por diante, caracterizando uma série de laços que se aninham dentro de outros laços.

Por fim, podemos utilizar um fluxograma para fixar a idéia que permeia os loops aninhados de maneira visual.

[](attachment:ListasELoops.png)

<center><b>Figura 3:</b> Fluxograma de um loop aninhado.</center><br>


Partindo para o código, para criarmos um loop aninhado, basta escrever um laço dentro de outro.
Utilizaremos o exemplo do tempo em um relógio para ilustrar:

```python
#aninhando os laços
for minutos in range(60):
    for segundos in range(60):
        print(minutos,"min",segundos,"s")
```

### 6.2 Erros comuns em loops


É muito importante dobrarmos a atenção quando trabalhamos com loops, pois muitos erros "bobos" podem acontecer e, às vezes, podem custar caro (em termos de memória e horas de depuração). Dentro desses, podemos citar:

- Esquecer de mudar o estado de uma condição de parada
- Esquecer a própria condição de parada
- Condição falsa desde o princípio

Esses erros podem nos levar a problemas simples como a não execução de um laço até erros mais complexos como o chamado __loop infinito__.

Podemos checar alguns dos comportamentos relacionados a tais erros nos exemplos abaixo:


```python
numero = 42

while numero < 5 :
    print("O numero é menor do que 5")
    numero += 1
```

O código dentro desse laço foi executado? Porquê?

```python
nome = "Novelli"

while nome == "Novelli":
    print("Achei esse nome bonito")
    
print("Saindo do laço")    
```

Como visto, o programa não vai parar de imprimir "Achei esse nome bonito" nunca, pois a condição do laço será sempre verdadeira. Chamamos situações como essas de **loops infinitos**.
