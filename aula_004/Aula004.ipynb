### 4.2 O Loop `while`

O loop `while`  uma estrutura de repetio na qual uma condio  verificada antes que os comandos dentro do loop sejam executados. Se a condio for verdadeira, um conjunto de comandos  executado e, caso contrrio, os comandos no so executados. Assim que os comandos forem executados, a condio  avaliada novamente e todo o processo se repete.

```python
x = 0           # Inicializando x

print("Entrando no lao")

while x < 5 :   # Enquanto x for menor do que 5 execute:
    print(x)    # Imprima o valor de x
    x = x + 1   # Adicione 1 ao valor de x. Tambm pode ser feito como x += 1
    
print("Samos do lao")
```

Um bom exerccio  olhar o cdigo e tentar prever o passo a passo das linhas e comandos que o computador vai executar. Tambm podemos combinar condies lgicas na chamada do `while`. Por exemplo, olhe o cdigo abaixo e pense quantos nmeros sero impressos na tela: 

```python
x = 0

while x < 100 and x != 20 :
    print("X est valendo " , x )
    x += 1

print("Samos do lao")
```

### 4.3 O Loop `for`

O loop `for` tem a funcionalidade geral muito parecida com a de um loop `while`, porm, ele permite simplificar alguns tipos de laos especficos e  mais comumente utilizado. O `for` exige a pela criao de um contador e a definio da condio de parada na sua chamada. Veremos algumas formas de como chamar esse lao.

Podemos definir o `for` e passar um contador (ou iterador) e um domnio (ou alcance), como no exemplo a seguir:

```python
for x in range(0,5):   # Para "x" no intervalo de 0  5 (no inclusivo, intervalo aberto  direita)
    print(x)           # Imprima "x"
```

A funo `range()` pode ter uma quantidade varivel de parmetros. Voc pode, por exemplo, dar apenas o inteiro para que o `for` pare de percorrer:

```python
for contador in range(12): # Note que se o incio no for dado, o contador comea pelo 0
    print(contador)
```

A funo `range` pode receber o incio e o fim, como no primeiro exemplo de lao `for`, ou tambm receber 3 parmetros que definem: o incio do lao, o fim e o quanto ele deve acrescentar no contador por cada iterao:

```python
for numero in range( 2 , 100, 5 ):
    print(numero)
```

Ento j vimos algumas vantagens para essa implementao do `for` comparado ao `while`. No precisamos criar o contador antes do lao e a funo `range()` j toma conta da iterao sozinha. Mais uma coisa legal que podemos fazer como o `for`  percorrer listas! Olhem s:

```python
baralho = ["s" , "2" , "3" , "4" , "5" , "6" , "7" , "8" , "9" , "J" , "Q" , "K"  ]

for carta in baralho: # Para cada carta na lista baralho
    print( carta )    # Imprima a carta
    
```

De forma bem simples e direta pegamos um contador e fazer ele passar por todos os itens de uma lista individualmente a cada iterao. Fantstico, no? E esse arranjo serve para qualquer tipo de lista.

### 4.4 Desvios de fluxo

Existem tambm algumas formas de "desviar" o fluxo de um lao. Dois deles so o `break` e o `continue`. Por exemplo, considere a situao na qual seja necessrio voltar ao comeo do lao sem terminar de executar todo o bloco de cdigo. Usaremos para isso o `continue`, como no exemplo a seguir:

```python
x = 0

while x < 12 :

    x += 1
    if ( x % 2 == 0 ):  #Fazendo um teste para saber a paridade de "x"
        continue        #Se "x" for par, volte para o incio do lao
    
    print(x)
```

E no caso do `break`, podemos interromper completamente um lao, sair dele e continuar abaixo do bloco de cdigo, como no exemplo a seguir:

```python
x = 0

while x < 99 :    #Lao de "x" enquanto "x" for menor que 99

    x += 1
    if x == 8 :   #Se "x" for igual a 8
        break     #Saia do lao
        
    print(x)

print("Samos do lao")
```

## 5. Loops aninhados

Os loops aninhados nada mais so que sequncias de loops definidos dentro de outros loops, ou seja, laos dentro de laos. Parece confuso, certo? Talvez um exemplo ajude a ilustrar melhor a ideia.<br>

Um exemplo que podemos citar est no nosso prprio calendrio/relgio. Nosso ano possui 12 meses e se repete de janeiro a dezembro, caracterizando um lao. Dentro do nosso ano, temos nossos meses e dentro destes, temos os dias (mais um lao). No obstante, ainda temos as horas, dentro dessas, os minutos, dentro desses, os segundos e assim por diante, caracterizando uma srie de laos que se aninham dentro de outros laos.

Por fim, podemos utilizar um fluxograma para fixar a idia que permeia os loops aninhados de maneira visual.

[](attachment:ListasELoops.png)

<center><b>Figura 3:</b> Fluxograma de um loop aninhado.</center><br>


Partindo para o cdigo, para criarmos um loop aninhado, basta escrever um lao dentro de outro.
Utilizaremos o exemplo do tempo em um relgio para ilustrar:

```python
#aninhando os laos
for minutos in range(60):
    for segundos in range(60):
        print(minutos,"min",segundos,"s")
```
